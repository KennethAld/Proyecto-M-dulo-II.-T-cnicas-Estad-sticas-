# 1. Introducción

En estadística y probabilidad, una **variable aleatoria** es una función que asigna un valor numérico a cada resultado posible de un experimento aleatorio. Estas variables son fundamentales para modelar fenómenos inciertos y analizar datos. Las variables aleatorias permiten describir de forma cuantitativa eventos como el número de llamadas recibidas por hora, la altura de una persona o el número de veces que se obtiene cara al lanzar una moneda.


# 2. Variables Aleatorias

Las variables aleatorias se dividen principalmente en dos tipos:

## 🎲 2.1.1 Variables Aleatorias Discretas
Toman un número finito o numerable de valores. Un ejemplo clásico es el número que sale al lanzar un dado:

- Valores posibles: $ \{1, 2, 3, 4, 5, 6\} $
- Probabilidad asociada a cada valor: $ P(X = x_i) $

**Ejemplo:**
Una variable $ X $ que representa el número de llamadas por minuto.

Fórmulas:
- Valor esperado: $ \mathbb{E}[X] = \sum_{i} x_i \cdot P(X = x_i) $
- Varianza: $ \text{Var}(X) = \sum_{i} (x_i - \mathbb{E}[X])^2 \cdot P(X = x_i) $



## 2.1.2 Variable aleatoria uniforme discreta

Una variable aleatoria $X$ tiene **distribución uniforme discreta** en el conjunto $\{x_{1},\dots,x_{n}\}$ si su función de masa de probabilidades está dada por:

$$
f_{X}(k)=\mathbb{P}(X=k)=\begin{cases}
\frac{1}{n} &\quad \text{si } k\in\{x_{1},\dots,x_{n}\} \\
0 &\quad \text{en otro caso}
\end{cases}
$$

**Notación:** $X\sim Unif(x_{1},\dots,x_{n})$.

Se llama **uniforme** porque cada uno de los posibles valores de $X$ tiene la misma probabilidad.

Si $X\sim Unif(x_{1},\dots,x_{n})$, se cumple:

$$
\mathbb{E}[X] = \sum_{k}k f_{X}(k) = \frac{x_n + x_1}{2},
$$

$$
\text{Var}(X) = \mathbb{E}[(X-\mathbb{E}[X])^{2}] = \sum_{k}(k-\mathbb{E}[X])^{2}f_{X}(k) = \frac{(x_n - x_1 + 1)^2 - 1}{12}.
$$

$\color{red}{\text{Ejercicio 1.}}$ Verifica que, en efecto, la esperanza y varianza de una variable aleatoria uniforme discreta están dadas como antes.

---

### **Ejemplo**

Se lleva a cabo una rifa donde los boletos están enumerados del $00$ al $99$. Si $Y$ es la variable aleatoria definida como el número del boleto ganador, entonces:

$$
\mathbb{P}(Y = k) = \begin{cases}
\frac{1}{100} &\quad \text{si } k = 0, 1, \dots, 99 \\
0 &\quad \text{en otro caso}
\end{cases}
$$

Definamos una nueva variable aleatoria $X = Y + 1$, que representa el **monto del premio** en pesos. Como $Y$ toma valores en $\{0, 1, \dots, 99\}$, entonces $X$ toma valores en $\{1, 2, \dots, 100\}$. Por lo tanto:

$$
f_X(k) = \begin{cases}
\frac{1}{100} &\quad \text{si } k = 1, 2, \dots, 100 \\
0 &\quad \text{en otro caso}
\end{cases}
$$

La esperanza y varianza de $X$ están dadas por:

$$
\mathbb{E}[X] = \sum_{k=1}^{100}k\left(\frac{1}{100}\right) = \frac{1}{100} \cdot \sum_{k=1}^{100}k = \frac{1}{100} \cdot \left[\frac{100(100 + 1)}{2}\right] = 50.5,
$$

$$
\text{Var}(X) = \mathbb{E}[X^{2}] - (\mathbb{E}[X])^{2} = \sum_{k=1}^{100}k^{2}\left(\frac{1}{100}\right) - (50.5)^{2} = 833.25.
$$

Supongamos que nos interesa calcular la probabilidad de que el premio sea mayor a \$80, entonces:

$$
\mathbb{P}(X > 80) = \sum_{k=81}^{100} \frac{1}{100} = \frac{20}{100} = 0.2.
$$

La función de distribución de $X$ está dada por:

$$
F_X(x) = \mathbb{P}(X \le x) = \begin{cases}
0 &\quad \text{si } x < 1 \\
\frac{x}{100} &\quad \text{si } x = 1, 2, \dots, 100 \\
1 &\quad \text{si } x > 100
\end{cases}
$$


```python
import matplotlib.pyplot as plt

x = range(1, 101)
probs = [1/100] * 100

plt.bar(x, probs, edgecolor='black')
plt.title('Distribución uniforme discreta sobre {1, 2, ..., 100}')
plt.xlabel('Valor de X')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 2.1.3 Variable Aleatoria Bernoulli

Una variable aleatoria $ X $ tiene distribución **Bernoulli** con parámetro $ p \in (0,1) $ si representa un experimento con **dos posibles resultados**: éxito y fracaso. Definimos:

- $ X = 1 $ si ocurre un éxito.
- $ X = 0 $ si ocurre un fracaso.

Las probabilidades asociadas son:

- $ \mathbb{P}(X = 1) = p $
- $ \mathbb{P}(X = 0) = 1 - p $

**Notación:** $ X \sim \text{Ber}(p) $.

#### **Interpretación práctica**

La distribución de Bernoulli es extremadamente útil en situaciones en las que solo se observan dos resultados posibles. Algunos ejemplos incluyen:

- El estado de una línea telefónica: **ocupada (1)** o **libre (0)**.
- La presencia o ausencia de una enfermedad en un paciente: **enfermo (1)** o **sano (0)**.
- La preferencia por un candidato en una elección: **a favor (1)** o **en contra (0)**.

#### **Función de masa de probabilidad**

La función de masa de probabilidad de $ X $ es:

$$
f_X(k) = \begin{cases}
p & \quad \text{si } k = 1 \\
1 - p & \quad \text{si } k = 0
\end{cases}
$$

#### **Esperanza y Varianza**

- La **esperanza** de $ X $ es:

$$
\mathbb{E}[X] = p
$$

- La **varianza** de $ X $ es:

$$
\text{Var}(X) = p(1 - p)
$$

#### **Ejemplo**

Supongamos que estamos modelando el **estado de un teléfono móvil** en un momento dado. Si $ X = 1 $ representa que el teléfono está **ocupado** y $ X = 0 $ significa que está **libre**, y sabemos que la probabilidad de que el teléfono esté ocupado en un momento dado es $ p = 0.7 $ .

Entonces, $ X \sim \text{Ber}(0.7) $, y la esperanza y varianza son:

$$
\mathbb{E}[X] = 0.7, \quad \text{Var}(X) = 0.7(1 - 0.7) = 0.21
$$


```python

import numpy as np
import matplotlib.pyplot as plt

# Parámetro p
p = 0.7

# Número de experimentos
n_experiments = 1000

# Simulación de la variable aleatoria Bernoulli
bernoulli_samples = np.random.binomial(1, p, n_experiments)

# Cálculo de la esperanza y la varianza simuladas
mean = np.mean(bernoulli_samples)
variance = np.var(bernoulli_samples)

print(f"Esperanza simulada: {mean}")
print(f"Varianza simulada: {variance}")

# Visualización de los resultados
plt.hist(bernoulli_samples, bins=2, edgecolor='black', alpha=0.7)
plt.title('Distribución de la variable aleatoria Bernoulli')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.xticks([0, 1], ['Fracaso (0)', 'Éxito (1)'])
plt.show()
```

### 2.1.4 Variable Aleatoria Binomial

Una variable aleatoria $ X $ tiene distribución binomial con parámetros $ n $ y $ p \in (0,1) $ si su función de densidad de masa de probabilidades está dada por:
$$
f_X(k) = \begin{cases} \binom{n}{k} p^k (1-p)^{n-k} & \quad \text{si } k = 0, 1, \dots, n \\ 0 & \quad \text{en otro caso} \end{cases} $$

donde $ n $ es un entero positivo y $ 0 < p < 1 $.

Esperanza y Varianza
La esperanza de $ X $ es:
$$
\mathbb{E}[X] = np $$

La varianza de $ X $ es:
$$
\text{Var}(X) = np(1-p) $$

Ejemplo
Supongamos que una moneda no justa se lanza 6 veces, y la probabilidad de que salga sol en cada lanzamiento es $ p = 0.3 $. Sea $ X $ el número de veces que cae sol en los 6 lanzamientos.

Entonces, $ X \sim \text{Binomial}(6, 0.3) $, y podemos calcular algunas probabilidades de interés.

Por ejemplo:
$$
\mathbb{P}(X = 2) = \binom{6}{2} (0.3)^2 (0.7)^4 = 0.3241 $$
$$
\mathbb{P}(X = 3) = \binom{6}{3} (0.3)^3 (0.7)^3 = 0.1852 $$

La probabilidad de que $ X $ esté en el intervalo $ 1 < X \leq 5 $ es:
$$
\mathbb{P}(1 < X \leq 5) = \mathbb{P}(X = 2) + \mathbb{P}(X = 3) + \mathbb{P}(X = 4) + \mathbb{P}(X = 5) $$

Calculamos esto:
$$
\mathbb{P}(X = 4) = \binom{6}{4} (0.3)^4 (0.7)^2 = 0.1323 $$
$$
\mathbb{P}(X = 5) = \binom{6}{5} (0.3)^5 (0.7)^1 = 0.027 $$

Así que:
$$
\mathbb{P}(1 < X \leq 5) = 0.3241 + 0.1852 + 0.1323 + 0.027 = 0.6686 $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución binomial
n = 6  # número de lanzamientos
p = 0.3  # probabilidad de éxito

# Número de experimentos
n_experiments = 1000

# Simulación de la variable aleatoria binomial
binomial_samples = np.random.binomial(n, p, n_experiments)

# Cálculo de la esperanza y la varianza simuladas
mean = np.mean(binomial_samples)
variance = np.var(binomial_samples)

print(f"Esperanza simulada: {mean}")
print(f"Varianza simulada: {variance}")

# Visualización de los resultados
plt.hist(binomial_samples, bins=np.arange(-0.5, n+1.5, 1), edgecolor='black', alpha=0.7)
plt.title('Distribución de la variable aleatoria Binomial')
plt.xlabel('Número de éxitos')
plt.ylabel('Frecuencia')
plt.xticks(np.arange(0, n+1, 1))
plt.show()
```

### 2.1.5 Variable Aleatoria Poisson

La distribución de Poisson es una distribución de probabilidad discreta que se utiliza para calcular la probabilidad de que ocurra un determinado número de eventos raros durante un intervalo dado (puede ser de tiempo, longitud, área, etc.).

Esta variable aleatoria toma los valores sobre el conjunto $ \{0, 1, 2, \dots \} $ y tiene un parámetro $ \lambda > 0 $ , el cual representa el número de veces que se espera que ocurra un evento durante el intervalo dado.

La función de masa de probabilidades de una variable aleatoria Poisson, $ X $ , está dada por:

$$
f_X(k) =
\begin{cases}
e^{-\lambda} \frac{\lambda^k}{k!} & \quad \text{si } k = 0, 1, 2, \dots \\
0 & \quad \text{en otro caso}
\end{cases}
$$

En efecto, es una función de masa de probabilidades debido a que:

$$
\sum_{k=0}^{\infty} e^{-\lambda} \frac{\lambda^k}{k!} = e^{-\lambda} \sum_{k=0}^{\infty} \frac{\lambda^k}{k!}
$$

$$
= e^{-\lambda} \underbrace{\left(1 + \lambda + \frac{\lambda^2}{2!} + \frac{\lambda^3}{3!} + \cdots \right)}_{\text{desarrollo de Taylor de } e^\lambda}
$$

$$
= e^{-\lambda} e^\lambda = 1
$$

#### **Esperanza y Varianza**

- La **esperanza** de $ X $ es:

$$
\mathbb{E}[X] = \lambda
$$

- La **varianza** de $ X $ es:

$$
\text{Var}(X) = \lambda
$$

#### **Ejemplo**

Supongamos que el número de autos que pasan por una caseta de peaje por hora sigue una distribución Poisson con parámetro $ \lambda = 5 $ ,  es decir, se espera que pasen en promedio 5 autos por hora.

La probabilidad de que pasen exactamente 3 autos en una hora es:

$$
\mathbb{P}(X = 3) = e^{-5} \frac{5^3}{3!} = 0.1404
$$

La probabilidad de que pasen más de 4 autos (es decir, \( X > 4 \)) es:

$$
\mathbb{P}(X > 4) = 1 - \mathbb{P}(X \leq 4) = 1 - \left( \mathbb{P}(X=0) + \mathbb{P}(X=1) + \mathbb{P}(X=2) + \mathbb{P}(X=3) + \mathbb{P}(X=4) \right)
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro lambda de la distribución Poisson
lambda_param = 5

# Número de experimentos
n_experiments = 1000

# Simulación de la variable aleatoria Poisson
poisson_samples = np.random.poisson(lambda_param, n_experiments)

# Cálculo de la esperanza y varianza simuladas
mean = np.mean(poisson_samples)
variance = np.var(poisson_samples)

print(f"Esperanza simulada: {mean}")
print(f"Varianza simulada: {variance}")

# Visualización de los resultados
plt.hist(poisson_samples, bins=np.arange(min(poisson_samples), max(poisson_samples) + 1), edgecolor='black', alpha=0.7)
plt.title('Distribución de la variable aleatoria Poisson')
plt.xlabel('Número de autos')
plt.ylabel('Frecuencia')
plt.xticks(np.arange(min(poisson_samples), max(poisson_samples) + 1, 1))
plt.show()
```

## Variable Aleatoria Geométrica

La variable aleatoria geométrica modela el número de **ensayos hasta obtener el primer éxito**, donde cada ensayo es independiente y tiene probabilidad de éxito $0 < p < 1$.

### Definición

Si $X \sim \text{Geo}(p)$, su función de masa de probabilidades es:

$$
f_X(k) =
\begin{cases}
p(1-p)^{k-1} & \text{si } k = 1, 2, \dots \\
0 & \text{en otro caso}
\end{cases}
$$

Esto representa el número de **ensayos necesarios** para obtener el primer éxito.

### Alternativa

Si se desea modelar el **número de fracasos antes del primer éxito**, entonces:

$$
f_X(k) =
\begin{cases}
p(1-p)^k & \text{si } k = 0, 1, 2, \dots \\
0 & \text{en otro caso}
\end{cases}
$$

### Esperanza y Varianza

- $\mathbb{E}[X] = \frac{1}{p}$
- $\text{Var}(X) = \frac{1 - p}{p^2}$

### Ejemplo

La probabilidad de que un cliente compre un seguro en una llamada es $p = 0.2$. La probabilidad de lograr la primera venta en la **tercera llamada** es:

$$
\mathbb{P}(X = 3) = 0.2 \cdot (1 - 0.2)^2 = 0.128
$$
---

## Variable Aleatoria Binomial Negativa

Modela el número de **ensayos necesarios para obtener $r$ éxitos** en ensayos independientes con probabilidad de éxito $0 < p < 1$.

### Definición

Si $X \sim \text{BN}(r, p)$, su función de masa de probabilidades es:

$$
f_X(n) =
\begin{cases}
\binom{n-1}{r-1} p^r (1 - p)^{n - r} & \text{si } n = r, r+1, \dots \\
0 & \text{en otro caso}
\end{cases}
$$

### Esperanza y Varianza

- $\mathbb{E}[X] = \frac{r}{p}$
- $\text{Var}(X) = \frac{r(1 - p)}{p^2}$

### Ejemplo

Se lanza una moneda con probabilidad de cara $p = 0.3$ hasta obtener 4 caras. La probabilidad de obtener la cuarta cara **en el séptimo lanzamiento** es:

$$
\mathbb{P}(X = 7) = \binom{6}{3} (0.3)^4 (0.7)^3 = 0.0595
$$
---

## Variable Aleatoria Hipergeométrica

Se utiliza cuando se seleccionan elementos **sin reemplazo** de una población finita.

### Definición

Sean $N$ elementos en total, de los cuales $m$ son considerados "éxito", y se extraen $n$ elementos sin reemplazo. Si $X \sim \text{Hiper}(n, N, m)$, su función de masa de probabilidades es:

$$
f_X(i) = \frac{\binom{m}{i} \binom{N - m}{n - i}}{\binom{N}{n}} \quad \text{para } i = 0, 1, \dots, n
$$

### Esperanza y Varianza

- $\mathbb{E}[X] = \frac{n m}{N}$
- $\text{Var}(X) = \frac{n m}{N} \left(1 - \frac{m}{N}\right) \left( \frac{N - n}{N - 1} \right)$

### Aproximación

Si $m, N \gg n$, con $p = \frac{m}{N}$, entonces:

$$
\mathbb{P}(X = i) \approx \binom{n}{i} p^i (1 - p)^{n - i}
$$

### Ejemplo

Una urna contiene 10 bolas, 4 rojas y 6 verdes. Si se extraen 3 bolas sin reemplazo, la probabilidad de obtener exactamente 2 bolas rojas es:

$$
\mathbb{P}(X = 2) = \frac{\binom{4}{2} \binom{6}{1}}{\binom{10}{3}} = \frac{36}{120} = 0.3
$$



## 📈 2.2 Variables Aleatorias Continuas
Pueden tomar cualquier valor dentro de un intervalo. Por ejemplo, la estatura de una persona:

- Se describe mediante una función de densidad $ f(x) $
- $ P(a \leq X \leq b) = \int_a^b f(x) \, dx $

Fórmulas:
- Valor esperado: $ \mathbb{E}[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx $
- Varianza: $ \text{Var}(X) = \int_{-\infty}^{\infty} (x - \mathbb{E}[X])^2 \cdot f(x) \, dx $


## 2.2.1 Distribución Uniforme continua

Sea $X$ una variable aleatoria continua que se distribuye de forma uniforme en el intervalo $[a, b]$, donde $a < b$. Esto significa que todos los valores dentro del intervalo son igualmente probables, y fuera de él, la probabilidad es cero.

Decimos que $X$ tiene una distribución uniforme continua si su función de densidad de probabilidad (f.d.p.) está dada por:

$$
f_X(x) =
\begin{cases}
\frac{1}{b - a}, & \text{si } x \in [a, b] \\
0, & \text{en otro caso}
\end{cases}
$$

Esta función cumple con las propiedades básicas de una función de densidad:

- $f_X(x) \geq 0$ para todo $x$.
- La integral sobre todo el espacio es 1:

$$
\int_{-\infty}^{\infty} f_X(x) \, dx = \int_a^b \frac{1}{b - a} \, dx = \frac{1}{b - a}(b - a) = 1
$$

### Esperanza y varianza

La esperanza y la varianza de una variable aleatoria $X \sim \text{Uniforme}(a, b)$ están dadas por:

- **Esperanza:**

$$
\mathbb{E}[X] = \frac{a + b}{2}
$$

- **Varianza:**

$$
\text{Var}(X) = \frac{(b - a)^2}{12}
$$

### Función de distribución acumulada (F.D.A.)

La función de distribución acumulada $F_X(x)$ está dada por:

$$
F_X(x) =
\begin{cases}
0, & \text{si } x < a \\
\frac{x - a}{b - a}, & \text{si } a \leq x \leq b \\
1, & \text{si } x > b
\end{cases}
$$

---

###  Ejemplo

Supongamos que el tiempo de espera (en minutos) para que pase un autobús está uniformemente distribuido entre 5 y 15 minutos. Entonces, si $X$ representa este tiempo de espera, tenemos que $X \sim \text{Uniforme}(5, 15)$.

- $f_X(x) = \frac{1}{15 - 5} = \frac{1}{10}$ para $x \in [5, 15]$.
- $\mathbb{E}[X] = \frac{5 + 15}{2} = 10$ minutos.
- $\text{Var}(X) = \frac{(15 - 5)^2}{12} = \frac{100}{12} \approx 8.33$

Calcular la probabilidad de que el autobús llegue entre los 8 y los 12 minutos:

$$
P(8 \leq X \leq 12) = \int_8^{12} \frac{1}{10} \, dx = \frac{12 - 8}{10} = 0.4
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parámetros de la distribución
a = 5  # límite inferior
b = 15  # límite superior

# Creamos una distribución uniforme en [a, b]
dist = uniform(loc=a, scale=b - a)

# Valores para graficar
x = np.linspace(a - 2, b + 2, 1000)
y = dist.pdf(x)

# Graficamos la función de densidad
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='f.d.p. Uniforme[5, 15]', color='blue')
plt.fill_between(x, y, where=(x >= 8) & (x <= 12), color='skyblue', alpha=0.5,
                 label='Área entre 8 y 12')
plt.title('Función de densidad de una Uniforme continua')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Cálculo de probabilidades
prob = dist.cdf(12) - dist.cdf(8)
print(f"La probabilidad de que X esté entre 8 y 12 es: {prob:.2f}")
```

## 2.2.2 Distribución Exponencial

Una variable aleatoria continua $X$ sigue una **Distribución Exponencial** con parámetro $\theta > 0$ si su función de densidad está dada por:

$$
f_X(x) = \begin{cases}
\theta e^{-\theta x} & \text{si } x \ge 0 \\
0 & \text{en otro caso}
\end{cases}
$$

Se denota $X \sim \text{Exp}(\theta)$. Esta distribución se utiliza frecuentemente para modelar **el tiempo entre eventos** en un proceso de Poisson.

La **función de distribución acumulada (FDA)** es:

$$
F_X(x) = \begin{cases}
1 - e^{-\theta x} & \text{si } x \ge 0 \\
0 & \text{en otro caso}
\end{cases}
$$

Las siguientes son propiedades importantes:
- Esperanza: $\mathbb{E}[X] = \frac{1}{\theta}$
- Varianza: $\text{Var}(X) = \frac{1}{\theta^2}$


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro theta
theta = 0.5

# Valores de x
x = np.linspace(0, 15, 1000)

# FDP y FDA
fdp = theta * np.exp(-theta * x)
fda = 1 - np.exp(-theta * x)

# Gráfica de la FDP
plt.figure(figsize=(8, 4))
plt.plot(x, fdp, label='FDP', color='purple')
plt.title("Función de Densidad de Probabilidad (Exponencial)")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.grid()
plt.legend()
plt.show()

# Gráfica de la FDA
plt.figure(figsize=(8, 4))
plt.plot(x, fda, label='FDA', color='green')
plt.title("Función de Distribución Acumulada (Exponencial)")
plt.xlabel("x")
plt.ylabel("Probabilidad acumulada")
plt.grid()
plt.legend()
plt.show()
```

### Ejemplo: Probabilidad entre dos valores

Supongamos $X \sim \text{Exp}(0.5)$. ¿Cuál es la probabilidad de que $X$ esté entre 2 y 5?

$$
\mathbb{P}(2 \le X \le 5) = F_X(5) - F_X(2) = \left(1 - e^{-0.5 \cdot 5}\right) - \left(1 - e^{-0.5 \cdot 2}\right)
$$


```python
# Cálculo usando FDA
from math import exp

p_2_5 = (1 - exp(-0.5 * 5)) - (1 - exp(-0.5 * 2))
print(f"P(2 <= X <= 5) = {p_2_5:.4f}")
```

### Aplicación: estimador de máxima verosimilitud (MLE)

Supongamos que tenemos 5 observaciones independientes de una variable $X \sim \text{Exp}(\theta)$, y el total observado es 14. La función de verosimilitud está dada por:

$$
L(\theta) = \theta^5 e^{-14 \theta}
$$

Entonces, la log-verosimilitud es:

$$
\log L(\theta) = 5 \log \theta - 14\theta
$$

A continuación derivamos esta función y resolvemos para encontrar el valor de $\theta$ que maximiza la verosimilitud.


```python
import sympy as sp

# Definir la variable simbólica para theta
theta = sp.Symbol('theta', positive=True)

# Log-verosimilitud
log_likelihood = 5 * sp.log(theta) - 14 * theta

# Derivar y resolver
log_likelihood_derivative = sp.diff(log_likelihood, theta)
theta_mle = sp.solve(log_likelihood_derivative, theta)

# Resultado
print(f"El MLE para theta es: {theta_mle[0]}")
```

## 2.2.3 Distribución Normal

Una variable aleatoria continua $X$ sigue una **Distribución Normal** con media $\mu \in \mathbb{R}$ y desviación estándar $\sigma > 0$ si su función de densidad está dada por:

$$
f_X(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left( -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 \right), \quad x \in \mathbb{R}
$$

Se denota $X \sim \mathcal{N}(\mu, \sigma^2)$. Esta distribución es fundamental en estadística y probabilidad debido al **Teorema Central del Límite** y su uso para modelar fenómenos naturales y sociales.

### Propiedades importantes
- Simétrica respecto a $\mu$.
- Esperanza: $\mathbb{E}[X] = \mu$
- Varianza: $\text{Var}(X) = \sigma^2$
- La curva tiene forma de campana, conocida como **curva gaussiana**.

Cuando $\mu = 0$ y $\sigma = 1$, se dice que $X$ tiene una **Distribución Normal Estándar**, denotada $X \sim \mathcal{N}(0,1)$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución normal
mu = 5
sigma = 2

# Valores de x
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Función de densidad
fdp = norm.pdf(x, mu, sigma)

# Graficar la FDP
plt.figure(figsize=(8, 4))
plt.plot(x, fdp, label=f"$\mathcal{{N}}({mu}, {sigma**2})$", color='darkred')
plt.title("Función de Densidad de Probabilidad (Normal)")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.grid()
plt.legend()
plt.show()
```

### Cálculo de probabilidades

Supongamos que $X \sim \mathcal{N}(5, 4)$, es decir, $\mu = 5$ y $\sigma = 2$.

#### ¿Cuál es la probabilidad de que $X$ esté entre 3 y 7?
Esto se calcula como:

$$
\mathbb{P}(3 \le X \le 7) = F_X(7) - F_X(3)
$$

donde $F_X(x)$ es la función de distribución acumulativa.


```python
# Cálculo de la probabilidad P(3 ≤ X ≤ 7)
p_3_7 = norm.cdf(7, mu, sigma) - norm.cdf(3, mu, sigma)
print(f"P(3 ≤ X ≤ 7) = {p_3_7:.4f}")
```

### Estandarización

Si $X \sim \mathcal{N}(\mu, \sigma^2)$, entonces la variable

$$
Z = \frac{X - \mu}{\sigma}
$$

sigue una distribución normal estándar: $Z \sim \mathcal{N}(0, 1)$. Esto permite usar las tablas de la normal estándar para calcular cualquier probabilidad.


## 2.2.4 Distribución Gamma

Una variable aleatoria continua $X$ tiene una **Distribución Gamma** con parámetros $\alpha > 0$ (forma) y $\theta > 0$ (escala) si su función de densidad está dada por:

$$
f_X(x) = \frac{1}{\Gamma(\alpha)\theta^\alpha} x^{\alpha - 1} e^{-x/\theta}, \quad x > 0
$$

donde $\Gamma(\alpha)$ es la función gamma, una generalización del factorial: $\Gamma(n) = (n-1)!$ para $n \in \mathbb{N}$.

Notación: $X \sim \text{Gamma}(\alpha, \theta)$

**Propiedades**:
- Esperanza: $\mathbb{E}[X] = \alpha \theta$
- Varianza: $\text{Var}(X) = \alpha \theta^2$

La distribución exponencial es un caso particular de la Gamma con $\alpha = 1$.

---

## 2.2.5 Distribución Beta

Una variable aleatoria continua $X$ sigue una **Distribución Beta** con parámetros $\alpha, \beta > 0$ si su función de densidad está dada por:

$$
f_X(x) = \frac{1}{B(\alpha, \beta)} x^{\alpha - 1}(1 - x)^{\beta - 1}, \quad 0 < x < 1
$$

donde $B(\alpha, \beta)$ es la función beta, que normaliza la densidad y se define como:

$$
B(\alpha, \beta) = \int_0^1 t^{\alpha - 1}(1 - t)^{\beta - 1} dt = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

Notación: $X \sim \text{Beta}(\alpha, \beta)$

**Propiedades**:
- Esperanza: $\mathbb{E}[X] = \frac{\alpha}{\alpha + \beta}$
- Varianza: $\text{Var}(X) = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$

La Beta se usa comúnmente en inferencia bayesiana y para modelar proporciones o probabilidades.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, beta

x_gamma = np.linspace(0, 20, 1000)
x_beta = np.linspace(0, 1, 1000)

# Parámetros para Gamma: (forma, escala)
gamma_params = [(1, 2), (2, 2), (5, 1)]

plt.figure(figsize=(12, 4))
for a, scale in gamma_params:
    plt.plot(x_gamma, gamma.pdf(x_gamma, a, scale=scale), label=fr'$\alpha={a}, \theta={scale}$')
plt.title("Distribución Gamma")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()

# Parámetros para Beta: (alpha, beta)
beta_params = [(0.5, 0.5), (2, 5), (5, 1)]

plt.figure(figsize=(12, 4))
for a, b in beta_params:
    plt.plot(x_beta, beta.pdf(x_beta, a, b), label=fr'$\alpha={a}, \beta={b}$')
plt.title("Distribución Beta")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```

# 2.3 Simulación de Variables Aleatorias

La simulación de variables aleatorias es un proceso mediante el cual generamos muestras aleatorias de una variable con una distribución dada. Este proceso es muy útil en situaciones donde obtener soluciones analíticas exactas no es posible o cuando se requiere realizar experimentos para verificar teorías. Existen diversos métodos para realizar simulaciones, y algunos de los más comunes son los siguientes:

## 2.3.1 Métodos de Simulación

### 1. Método de la Transformada Inversa

El **método de la transformada inversa** es uno de los más comunes y sencillos. Este método se utiliza cuando conocemos la función de distribución acumulativa (FDA) de una variable aleatoria $X$ y podemos invertirla. El procedimiento es el siguiente:

1. **Generar un número aleatorio $u$**: Generamos un número aleatorio $u \sim U(0,1)$.
2. **Aplicar la inversa de la FDA**: Si conocemos la FDA de la variable aleatoria $X$, denotada por $F_X(x)$, entonces la variable aleatoria generada es:
   $$ X = F_X^{-1}(u) $$

#### Ejemplo con la Distribución Exponencial

Supongamos que queremos generar una variable aleatoria $X \sim \text{Exp}(\lambda)$, cuya FDA es:
$$ F_X(x) = 1 - e^{-\lambda x} $$

La inversa de esta FDA es:
$$ F_X^{-1}(u) = -\frac{1}{\lambda} \ln(1 - u) $$

Dado que $u \sim U(0, 1)$, podemos generar muestras de $X$ como:
$$ X = -\frac{1}{\lambda} \ln(u) $$

### 2. Método de Aceptación-Rechazo

El **método de aceptación-rechazo** es más general y no requiere la inversión de la FDA. En lugar de eso, se utiliza una función de densidad de probabilidad (FDP) que se pueda comparar fácilmente con una distribución conocida. El algoritmo es el siguiente:

1. **Generar un valor $y$ de una distribución candidata**: Seleccionamos una distribución candidata fácil de simular, de la cual generamos un valor $y$.
2. **Generar un valor de aceptación $u$**: Generamos un número aleatorio $u \sim U(0,1)$.
3. **Aceptar o rechazar**: Si $u \leq \frac{f_X(y)}{M f_Y(y)}$, donde $f_X(y)$ es la FDP de la distribución objetivo y $f_Y(y)$ es la FDP de la distribución candidata, aceptamos $y$ como una muestra de $X$. Si no, rechazamos $y$ y repetimos el proceso.

Este método es útil cuando la función de densidad objetivo no es fácilmente invertible.

### 3. Método de Monte Carlo

El **método de Monte Carlo** no está necesariamente relacionado con la generación directa de variables aleatorias, sino con la estimación de funciones matemáticas mediante simulaciones. Se utiliza cuando no es posible obtener una solución cerrada de un problema, pero se puede estimar mediante la simulación.

El proceso básico es:

1. **Generar muestras aleatorias** de variables con distribuciones conocidas.
2. **Realizar experimentos** o cálculos con esas muestras para estimar el valor de una función o un parámetro.

Por ejemplo, se podría utilizar el método de Monte Carlo para estimar la integral de una función o para aproximar el valor de una expectativa.

### 4. Método de la Serie de Taylor (o de Aproximación Estocástica)

El **método de la serie de Taylor** se usa cuando se tiene una función de distribución difícil de simular. En este método, se emplea una expansión de Taylor de una función aleatoria para aproximar su valor esperado o su distribución. Es útil para generar muestras de distribuciones que se relacionan con distribuciones conocidas, mediante la adición de perturbaciones aleatorias.

Este método es más común en aplicaciones avanzadas, como en la simulación de procesos estocásticos.

## Elección del Método de Simulación

La elección de un método depende de varios factores:
- **Facilidad de simulación**: Algunas distribuciones son más fáciles de simular que otras (por ejemplo, la distribución uniforme).
- **Requerimientos de precisión**: Si se necesita una simulación precisa, se puede preferir el método de la transformada inversa, siempre que sea posible.
- **Disponibilidad de funciones**: Algunos lenguajes de programación tienen bibliotecas para generar distribuciones específicas de manera eficiente.

En resumen, los métodos de simulación permiten generar muestras aleatorias de diferentes distribuciones, facilitando el análisis de fenómenos estocásticos en diversas aplicaciones.


## 2.3.2 Importancia de la Simulación en la Estadística Aplicada

La **simulación** es una herramienta fundamental en la estadística aplicada, especialmente cuando se trata de modelar fenómenos complejos o realizar experimentos donde las soluciones analíticas no son factibles. Algunos de los principales motivos de su importancia son:

En muchas situaciones, las distribuciones no tienen una solución cerrada, lo que hace que los métodos analíticos tradicionales sean poco prácticos. La simulación es útil para obtener una aproximación empírica de las distribuciones de variables aleatorias, especialmente cuando no se conocen explícitamente sus funciones de densidad o acumulativas. A través de la simulación, podemos generar grandes muestras de variables aleatorias y obtener estimaciones de sus parámetros, como la media, la varianza y otras características.

### 3. **Estimación de Parámetros y Modelos**
Uno de los usos más comunes de la simulación en estadística aplicada es la estimación de parámetros en modelos estadísticos. Técnicas como el **Método de Monte Carlo** o el **Método de la Transformada Inversa** permiten obtener estimaciones precisas de parámetros que podrían ser difíciles de calcular directamente, como en la estimación de probabilidades o la evaluación de integrales en modelos estadísticos complejos.

### 4. **Análisis de Riesgos y Modelado Estocástico**
En la estadística aplicada, muchas veces estamos interesados en modelar fenómenos estocásticos, como la evolución de los precios en los mercados financieros o el análisis de riesgos en sistemas de ingeniería. La simulación permite generar escenarios aleatorios que nos dan una comprensión más profunda de los comportamientos y riesgos inherentes a dichos sistemas.

### 5. **Validación de Modelos**
Los modelos matemáticos y estadísticos son frecuentemente probados y validados mediante simulaciones. En lugar de aplicar el modelo a datos reales, lo que puede ser costoso o inviable, se puede simular el comportamiento de un sistema para comprobar si el modelo es adecuado para predecir su comportamiento. Este enfoque es particularmente útil en áreas como la ingeniería, la biología computacional y las ciencias sociales.

### 6. **Mejora en la Toma de Decisiones**
La simulación también es clave en la toma de decisiones bajo incertidumbre. Por ejemplo, en la optimización de inventarios, la simulación puede ayudar a evaluar diferentes estrategias de gestión de inventarios para encontrar la que minimice los costos o maximice la eficiencia.

### 7. **Generación de Datos Sintéticos**
En muchas ocasiones, no se dispone de datos reales suficientes o adecuados para realizar análisis. La simulación permite generar **datos sintéticos** que imitan el comportamiento de los datos reales, lo que permite probar algoritmos y modelos estadísticos sin necesidad de contar con una gran cantidad de datos reales.

### 8. **Educación y Formación**
La simulación es una herramienta excelente para la enseñanza de la estadística aplicada. Permite a los estudiantes visualizar cómo funcionan las distribuciones, los modelos y los algoritmos en situaciones reales, facilitando la comprensión de conceptos teóricos complejos.



## Ejemplo 1. Simulación de Monte Carlo para estimar la integral de una función

```python
import numpy as np

# Función a integrar
def f(x):
    return np.sin(x)  # Ejemplo con la función seno

# Método de Monte Carlo para estimar la integral de f(x) desde a hasta b
def monte_carlo_integration(f, a, b, n):
    x = np.random.uniform(a, b, n)  # Generar n muestras uniformemente distribuidas entre a y b
    y = f(x)  # Evaluar la función en esas muestras
    return (b - a) * np.mean(y)  # Aproximación a la integral

# Parámetros
a, b = 0, np.pi  # Intervalo de integración
n = 100000  # Número de muestras

# Estimación de la integral
integral_estimada = monte_carlo_integration(f, a, b, n)
print(f"Estimación de la integral de sin(x) entre 0 y pi: {integral_estimada:.4f}")
```

## Ejemplo 2. Simulación para analisis de riesgo en un sistema de inventarios

```python
import matplotlib.pyplot as plt

# Parámetros de la simulación
mean_demand = 50  # Media de la demanda diaria
std_demand = 10   # Desviación estándar de la demanda diaria
stock_level = 60  # Nivel de inventario inicial

# Número de simulaciones
n_simulations = 1000

# Función para simular el comportamiento de la demanda
def simulate_inventory(n_simulations, mean_demand, std_demand, stock_level):
    days = np.random.normal(mean_demand, std_demand, n_simulations)  # Demanda diaria simulada
    shortages = np.sum(days > stock_level)  # Número de días con desabastecimiento
    return shortages / n_simulations  # Proporción de días con desabastecimiento

# Simulación
risk = simulate_inventory(n_simulations, mean_demand, std_demand, stock_level)
print(f"Riesgo de desabastecimiento (proporción de días): {risk:.4f}")

# Graficar la distribución de las demandas simuladas
plt.hist(np.random.normal(mean_demand, std_demand, n_simulations), bins=30, color='blue', alpha=0.7)
plt.title('Distribución Simulada de Demanda Diaria')
plt.xlabel('Demanda')
plt.ylabel('Frecuencia')
plt.show()
```

# 3.1 Función Generadora de Momentos (FGM)

### Definición

La **Función Generadora de Momentos (FGM)** de una variable aleatoria \( X \) es una función matemática que se utiliza para obtener los momentos (es decir, la esperanza, la varianza, etc.) de la distribución de \( X \). Está definida como el valor esperado de la exponencial de \( tX \), es decir:

$$
M_X(t) = \mathbb{E}[e^{tX}]
$$

Donde:
- \( M_X(t) \) es la FGM de la variable aleatoria \( X \),
- \( t \) es un parámetro real, y
- \( \mathbb{E}[e^{tX}] \) es el valor esperado de la función \( e^{tX} \).

### Propiedades de la Función Generadora de Momentos

1. **Existencia**:
   La FGM de \( X \) existe en un intervalo abierto que contiene \( t = 0 \) (siempre que \( X \) tenga momentos finitos).

2. **Momento de orden \( n \)**:
   El \( n \)-ésimo momento de \( X \), \( \mathbb{E}[X^n] \), puede obtenerse a partir de la FGM derivando la función generadora de momentos \( n \) veces con respecto a \( t \) y evaluando en \( t = 0 \):
   $$
   \mathbb{E}[X^n] = \frac{d^n M_X(t)}{dt^n} \Big|_{t=0}
   $$

3. **Propiedad de la linealidad**:
   Si \( X \) y \( Y \) son variables aleatorias, la FGM de la suma \( X + Y \) es el producto de las FGMs de \( X \) y \( Y \), siempre que \( X \) y \( Y \) sean independientes:
   $$
   M_{X+Y}(t) = M_X(t) \cdot M_Y(t)
   $$

4. **Función generadora de momentos de variables transformadas**:
   Si \( Y = aX + b \), entonces la FGM de \( Y \) es:
   $$
   M_Y(t) = e^{bt} M_X(at)
   $$

5. **Determinar la distribución**:
   En algunos casos, la FGM es única para una distribución, lo que significa que la FGM puede ser utilizada para identificar completamente la distribución de una variable aleatoria.

### Aplicaciones de la Función Generadora de Momentos

- **Cálculo de momentos**: La FGM es útil para calcular momentos de una distribución sin necesidad de recurrir a la definición directa de \( \mathbb{E}[X^n] \).
  
- **Teoría de colas**: En procesos estocásticos, como las colas de espera, la FGM se utiliza para modelar el comportamiento de las llegadas de clientes y las esperas en sistemas de servicio.
  
- **Distribuciones compuestas**: La FGM es útil para obtener la distribución de la suma de variables aleatorias, especialmente cuando estas variables son independientes.

- **Cálculo de la varianza y la covarianza**: La FGM es útil para obtener varianzas y covarianzas en distribuciones complejas o para obtener los momentos de distribución de una suma de variables aleatorias.



### Ejemplo: Visualización de la FGM para una distribución exponencial

Consideremos una variable aleatoria $ X \sim \text{Exponencial}(\lambda) $ La FGM de $ X $ es:

$$
M_X(t) = \mathbb{E}[e^{tX}] = \frac{\lambda}{\lambda - t}, \quad \text{para } t < \lambda
$$

Esta función está definida solo para $ t < \lambda $ ya que fuera de ese rango la esperanza no converge.

A continuación, graficamos esta función para visualizar cómo se comporta.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de la distribución exponencial
lambd = 2  # lambda > 0

# Rango de t para el que M_X(t) está definido: t < lambda
t_values = np.linspace(-1, lambd - 0.01, 500)
mgf_values = lambd / (lambd - t_values)

plt.figure(figsize=(8, 5))
plt.plot(t_values, mgf_values, label=r'$M_X(t) = \frac{\lambda}{\lambda - t}$', color='purple')
plt.axvline(lambd, color='red', linestyle='--', label=r'Límite $t = \lambda$')
plt.title('Función Generadora de Momentos (FGM) - Distribución Exponencial')
plt.xlabel('t')
plt.ylabel(r'$M_X(t)$')
plt.ylim(0, 10)
plt.grid(True)
plt.legend()
plt.show()
```

## 3.2 Función Característica (FC)

### Definición

La **función característica** de una variable aleatoria $ X $ es la función compleja definida por:

$$
\varphi_X(t) = \mathbb{E}[e^{itX}], \quad t \in \mathbb{R}
$$

donde $ i $ es la unidad imaginaria, es decir, $ i^2 = -1 $

La función característica **siempre existe** (a diferencia de la FGM) y determina completamente la distribución de una variable aleatoria.

---

### Aplicaciones y ventajas

- Permite derivar momentos de una variable aleatoria (cuando existen).
- Útil para estudiar **convergencia en distribución** (teorema central del límite).
- La función característica de la **suma de variables independientes** es el producto de sus funciones características.
- Permite derivar distribuciones límite en procesos estocásticos.
- A diferencia de la FGM, **existe para todas las variables aleatorias**.

---

### Ejemplo: Función característica de la distribución normal

Sea $ X \sim \mathcal{N}(\mu, \sigma^2) $ su función característica es:

$$
\varphi_X(t) = e^{it\mu - \frac{1}{2} \sigma^2 t^2}
$$

Vamos a graficar la parte real e imaginaria de $ \varphi_X(t) $ para distintos valores de $ t $



```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la normal
mu = 0.5
sigma = 1

# Función característica de la normal
def phi_X(t, mu=mu, sigma=sigma):
    return np.exp(1j * t * mu - 0.5 * (sigma**2) * t**2)

# Valores de t
t_vals = np.linspace(-10, 10, 500)
phi_vals = phi_X(t_vals)

# Graficar parte real e imaginaria
plt.figure(figsize=(10, 5))
plt.plot(t_vals, np.real(phi_vals), label='Parte real', color='blue')
plt.plot(t_vals, np.imag(phi_vals), label='Parte imaginaria', color='orange')
plt.title('Función Característica de una Variable Aleatoria Normal')
plt.xlabel('t')
plt.ylabel(r'$\varphi_X(t)$')
plt.axhline(0, color='black', linestyle='--', alpha=0.6)
plt.legend()
plt.grid(True)
plt.show()
```

## 4. Teoremas Fundamentales

### 4.1 Ley de los Grandes Números (LGN)

La **Ley de los Grandes Números** es un resultado esencial de la teoría de la probabilidad. Describe el comportamiento del promedio de una secuencia de variables aleatorias conforme el tamaño de la muestra crece. Tiene dos versiones principales: **forma débil** y **forma fuerte**.

---

###  Enunciado (Forma Débil)

Sea ${X_1, X_2, \dots, X_n}$ una sucesión de variables aleatorias i.i.d. (independientes e idénticamente distribuidas) con esperanza finita $\mu = \mathbb{E}[X_i]$. Entonces, el promedio muestral:

$$
\bar{X}_n = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

converge en probabilidad al valor esperado $\mu$, es decir:

$$
\bar{X}_n \xrightarrow{p} \mu \quad \text{cuando} \quad n \to \infty
$$

---

###  Enunciado (Forma Fuerte)

Bajo las mismas condiciones, si además $\mathbb{E}[|X_i|] < \infty$, entonces:

$$
\bar{X}_n \xrightarrow{c.s.} \mu \quad \text{cuando} \quad n \to \infty
$$

Esto significa que $\bar{X}_n$ converge **casi seguramente** a $\mu$ (con probabilidad 1), lo cual es una forma más fuerte de convergencia.

---

###  Implicaciones

- Permite justificar el uso del promedio muestral $\bar{X}_n$ como estimador del valor esperado poblacional $\mu$.
- Fundamenta el uso de simulaciones en estadística aplicada.
- Garantiza que con suficientes observaciones, el comportamiento aleatorio se estabiliza.
- Es la base de la inferencia estadística clásica.


```python
import numpy as np
import matplotlib.pyplot as plt

# Fijar la semilla para reproducibilidad
np.random.seed(0)

# Parámetros
n = 10000  # número de observaciones
mu = 5     # valor esperado verdadero
sigma = 2  # desviación estándar

# Generar variables aleatorias normales
X = np.random.normal(loc=mu, scale=sigma, size=n)

# Promedios muestrales acumulados
media_muestral = np.cumsum(X) / np.arange(1, n + 1)

# Gráfica
plt.figure(figsize=(10, 5))
plt.plot(media_muestral, label='Media muestral acumulada')
plt.axhline(y=mu, color='red', linestyle='--', label=r'$\mu = 5$')
plt.title("Ley de los Grandes Números (Simulación)")
plt.xlabel("Número de observaciones")
plt.ylabel("Media muestral")
plt.grid(True)
plt.legend()
plt.show()
```

## 4.2 Teorema Central del Límite (TCL)

El **Teorema Central del Límite** es uno de los resultados más importantes de la estadística y la probabilidad. Establece que, bajo ciertas condiciones, el comportamiento de la suma (o promedio) de variables aleatorias puede aproximarse por una distribución normal.

---

###  Enunciado

Sea ${X_1, X_2, \dots, X_n}$ una secuencia de variables aleatorias i.i.d. con valor esperado $\mu = \mathbb{E}[X_i]$ y varianza finita $\sigma^2 = \text{Var}(X_i)$. Entonces, al aumentar el tamaño de la muestra $n$, la variable aleatoria estandarizada:

$$
Z_n = \frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}} = \frac{1}{\sqrt{n}} \sum_{i=1}^{n} \frac{X_i - \mu}{\sigma}
$$

converge en distribución a una normal estándar:

$$
Z_n \xrightarrow{d} N(0, 1)
$$

---

###  Aplicación práctica

El **TCL** permite usar la distribución normal como aproximación en problemas donde se trabaja con sumas o promedios de muchas variables aleatorias, incluso si las variables originales no son normales.

**Aplicaciones clave:**

- Intervalos de confianza y pruebas de hipótesis.
- Simulación estadística.
- Control de calidad.
- Análisis de grandes bases de datos.

###  Relación con la distribución normal

El TCL explica por qué la normal aparece con tanta frecuencia en estadísticas muestrales. Aunque la población no sea normal, el promedio muestral se aproxima a una distribución normal cuando $n$ es grande.

Esto permite usar la normal como aproximación para otras distribuciones, por ejemplo:

- Binomial con $n$ grande.
- Poisson con $\lambda$ grande.
- Cualquier distribución con media y varianza finitas.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Fijar semilla para reproducibilidad
np.random.seed(42)

# Parámetros
n = 30          # tamaño de muestra
repeticiones = 10000
lambda_pois = 4  # parámetro de una distribución Poisson

# Simular muchas muestras de promedio de Poisson
medias = [np.mean(np.random.poisson(lam=lambda_pois, size=n)) for _ in range(repeticiones)]

# Gráfica del histograma de los promedios
plt.figure(figsize=(10,5))
plt.hist(medias, bins=50, density=True, alpha=0.6, color='skyblue', label='Distribución empírica')

# Superponer la normal teórica (TCL)
mu = lambda_pois
sigma = np.sqrt(lambda_pois / n)
x = np.linspace(min(medias), max(medias), 200)
plt.plot(x, norm.pdf(x, loc=mu, scale=sigma), 'r--', label='Normal aprox. (TCL)')

plt.title("Teorema Central del Límite con muestras Poisson")
plt.xlabel("Promedio muestral")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```
