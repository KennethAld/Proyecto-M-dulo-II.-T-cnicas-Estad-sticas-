# 1. Introducci칩n

En estad칤stica y probabilidad, una **variable aleatoria** es una funci칩n que asigna un valor num칠rico a cada resultado posible de un experimento aleatorio. Estas variables son fundamentales para modelar fen칩menos inciertos y analizar datos. Las variables aleatorias permiten describir de forma cuantitativa eventos como el n칰mero de llamadas recibidas por hora, la altura de una persona o el n칰mero de veces que se obtiene cara al lanzar una moneda.


# 2. Variables Aleatorias

Las variables aleatorias se dividen principalmente en dos tipos:

## 游 2.1.1 Variables Aleatorias Discretas
Toman un n칰mero finito o numerable de valores. Un ejemplo cl치sico es el n칰mero que sale al lanzar un dado:

- Valores posibles: $ \{1, 2, 3, 4, 5, 6\} $
- Probabilidad asociada a cada valor: $ P(X = x_i) $

**Ejemplo:**
Una variable $ X $ que representa el n칰mero de llamadas por minuto.

F칩rmulas:
- Valor esperado: $ \mathbb{E}[X] = \sum_{i} x_i \cdot P(X = x_i) $
- Varianza: $ \text{Var}(X) = \sum_{i} (x_i - \mathbb{E}[X])^2 \cdot P(X = x_i) $



## 2.1.2 Variable aleatoria uniforme discreta

Una variable aleatoria $X$ tiene **distribuci칩n uniforme discreta** en el conjunto $\{x_{1},\dots,x_{n}\}$ si su funci칩n de masa de probabilidades est치 dada por:

$$
f_{X}(k)=\mathbb{P}(X=k)=\begin{cases}
\frac{1}{n} &\quad \text{si } k\in\{x_{1},\dots,x_{n}\} \\
0 &\quad \text{en otro caso}
\end{cases}
$$

**Notaci칩n:** $X\sim Unif(x_{1},\dots,x_{n})$.

Se llama **uniforme** porque cada uno de los posibles valores de $X$ tiene la misma probabilidad.

Si $X\sim Unif(x_{1},\dots,x_{n})$, se cumple:

$$
\mathbb{E}[X] = \sum_{k}k f_{X}(k) = \frac{x_n + x_1}{2},
$$

$$
\text{Var}(X) = \mathbb{E}[(X-\mathbb{E}[X])^{2}] = \sum_{k}(k-\mathbb{E}[X])^{2}f_{X}(k) = \frac{(x_n - x_1 + 1)^2 - 1}{12}.
$$

$\color{red}{\text{Ejercicio 1.}}$ Verifica que, en efecto, la esperanza y varianza de una variable aleatoria uniforme discreta est치n dadas como antes.

---

### **Ejemplo**

Se lleva a cabo una rifa donde los boletos est치n enumerados del $00$ al $99$. Si $Y$ es la variable aleatoria definida como el n칰mero del boleto ganador, entonces:

$$
\mathbb{P}(Y = k) = \begin{cases}
\frac{1}{100} &\quad \text{si } k = 0, 1, \dots, 99 \\
0 &\quad \text{en otro caso}
\end{cases}
$$

Definamos una nueva variable aleatoria $X = Y + 1$, que representa el **monto del premio** en pesos. Como $Y$ toma valores en $\{0, 1, \dots, 99\}$, entonces $X$ toma valores en $\{1, 2, \dots, 100\}$. Por lo tanto:

$$
f_X(k) = \begin{cases}
\frac{1}{100} &\quad \text{si } k = 1, 2, \dots, 100 \\
0 &\quad \text{en otro caso}
\end{cases}
$$

La esperanza y varianza de $X$ est치n dadas por:

$$
\mathbb{E}[X] = \sum_{k=1}^{100}k\left(\frac{1}{100}\right) = \frac{1}{100} \cdot \sum_{k=1}^{100}k = \frac{1}{100} \cdot \left[\frac{100(100 + 1)}{2}\right] = 50.5,
$$

$$
\text{Var}(X) = \mathbb{E}[X^{2}] - (\mathbb{E}[X])^{2} = \sum_{k=1}^{100}k^{2}\left(\frac{1}{100}\right) - (50.5)^{2} = 833.25.
$$

Supongamos que nos interesa calcular la probabilidad de que el premio sea mayor a \$80, entonces:

$$
\mathbb{P}(X > 80) = \sum_{k=81}^{100} \frac{1}{100} = \frac{20}{100} = 0.2.
$$

La funci칩n de distribuci칩n de $X$ est치 dada por:

$$
F_X(x) = \mathbb{P}(X \le x) = \begin{cases}
0 &\quad \text{si } x < 1 \\
\frac{x}{100} &\quad \text{si } x = 1, 2, \dots, 100 \\
1 &\quad \text{si } x > 100
\end{cases}
$$


```python
import matplotlib.pyplot as plt

x = range(1, 101)
probs = [1/100] * 100

plt.bar(x, probs, edgecolor='black')
plt.title('Distribuci칩n uniforme discreta sobre {1, 2, ..., 100}')
plt.xlabel('Valor de X')
plt.ylabel('Probabilidad')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 2.1.3 Variable Aleatoria Bernoulli

Una variable aleatoria $ X $ tiene distribuci칩n **Bernoulli** con par치metro $ p \in (0,1) $ si representa un experimento con **dos posibles resultados**: 칠xito y fracaso. Definimos:

- $ X = 1 $ si ocurre un 칠xito.
- $ X = 0 $ si ocurre un fracaso.

Las probabilidades asociadas son:

- $ \mathbb{P}(X = 1) = p $
- $ \mathbb{P}(X = 0) = 1 - p $

**Notaci칩n:** $ X \sim \text{Ber}(p) $.

#### **Interpretaci칩n pr치ctica**

La distribuci칩n de Bernoulli es extremadamente 칰til en situaciones en las que solo se observan dos resultados posibles. Algunos ejemplos incluyen:

- El estado de una l칤nea telef칩nica: **ocupada (1)** o **libre (0)**.
- La presencia o ausencia de una enfermedad en un paciente: **enfermo (1)** o **sano (0)**.
- La preferencia por un candidato en una elecci칩n: **a favor (1)** o **en contra (0)**.

#### **Funci칩n de masa de probabilidad**

La funci칩n de masa de probabilidad de $ X $ es:

$$
f_X(k) = \begin{cases}
p & \quad \text{si } k = 1 \\
1 - p & \quad \text{si } k = 0
\end{cases}
$$

#### **Esperanza y Varianza**

- La **esperanza** de $ X $ es:

$$
\mathbb{E}[X] = p
$$

- La **varianza** de $ X $ es:

$$
\text{Var}(X) = p(1 - p)
$$

#### **Ejemplo**

Supongamos que estamos modelando el **estado de un tel칠fono m칩vil** en un momento dado. Si $ X = 1 $ representa que el tel칠fono est치 **ocupado** y $ X = 0 $ significa que est치 **libre**, y sabemos que la probabilidad de que el tel칠fono est칠 ocupado en un momento dado es $ p = 0.7 $ .

Entonces, $ X \sim \text{Ber}(0.7) $, y la esperanza y varianza son:

$$
\mathbb{E}[X] = 0.7, \quad \text{Var}(X) = 0.7(1 - 0.7) = 0.21
$$


```python

import numpy as np
import matplotlib.pyplot as plt

# Par치metro p
p = 0.7

# N칰mero de experimentos
n_experiments = 1000

# Simulaci칩n de la variable aleatoria Bernoulli
bernoulli_samples = np.random.binomial(1, p, n_experiments)

# C치lculo de la esperanza y la varianza simuladas
mean = np.mean(bernoulli_samples)
variance = np.var(bernoulli_samples)

print(f"Esperanza simulada: {mean}")
print(f"Varianza simulada: {variance}")

# Visualizaci칩n de los resultados
plt.hist(bernoulli_samples, bins=2, edgecolor='black', alpha=0.7)
plt.title('Distribuci칩n de la variable aleatoria Bernoulli')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.xticks([0, 1], ['Fracaso (0)', '칄xito (1)'])
plt.show()
```

### 2.1.4 Variable Aleatoria Binomial

Una variable aleatoria $ X $ tiene distribuci칩n binomial con par치metros $ n $ y $ p \in (0,1) $ si su funci칩n de densidad de masa de probabilidades est치 dada por:
$$
f_X(k) = \begin{cases} \binom{n}{k} p^k (1-p)^{n-k} & \quad \text{si } k = 0, 1, \dots, n \\ 0 & \quad \text{en otro caso} \end{cases} $$

donde $ n $ es un entero positivo y $ 0 < p < 1 $.

Esperanza y Varianza
La esperanza de $ X $ es:
$$
\mathbb{E}[X] = np $$

La varianza de $ X $ es:
$$
\text{Var}(X) = np(1-p) $$

Ejemplo
Supongamos que una moneda no justa se lanza 6 veces, y la probabilidad de que salga sol en cada lanzamiento es $ p = 0.3 $. Sea $ X $ el n칰mero de veces que cae sol en los 6 lanzamientos.

Entonces, $ X \sim \text{Binomial}(6, 0.3) $, y podemos calcular algunas probabilidades de inter칠s.

Por ejemplo:
$$
\mathbb{P}(X = 2) = \binom{6}{2} (0.3)^2 (0.7)^4 = 0.3241 $$
$$
\mathbb{P}(X = 3) = \binom{6}{3} (0.3)^3 (0.7)^3 = 0.1852 $$

La probabilidad de que $ X $ est칠 en el intervalo $ 1 < X \leq 5 $ es:
$$
\mathbb{P}(1 < X \leq 5) = \mathbb{P}(X = 2) + \mathbb{P}(X = 3) + \mathbb{P}(X = 4) + \mathbb{P}(X = 5) $$

Calculamos esto:
$$
\mathbb{P}(X = 4) = \binom{6}{4} (0.3)^4 (0.7)^2 = 0.1323 $$
$$
\mathbb{P}(X = 5) = \binom{6}{5} (0.3)^5 (0.7)^1 = 0.027 $$

As칤 que:
$$
\mathbb{P}(1 < X \leq 5) = 0.3241 + 0.1852 + 0.1323 + 0.027 = 0.6686 $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metros de la distribuci칩n binomial
n = 6  # n칰mero de lanzamientos
p = 0.3  # probabilidad de 칠xito

# N칰mero de experimentos
n_experiments = 1000

# Simulaci칩n de la variable aleatoria binomial
binomial_samples = np.random.binomial(n, p, n_experiments)

# C치lculo de la esperanza y la varianza simuladas
mean = np.mean(binomial_samples)
variance = np.var(binomial_samples)

print(f"Esperanza simulada: {mean}")
print(f"Varianza simulada: {variance}")

# Visualizaci칩n de los resultados
plt.hist(binomial_samples, bins=np.arange(-0.5, n+1.5, 1), edgecolor='black', alpha=0.7)
plt.title('Distribuci칩n de la variable aleatoria Binomial')
plt.xlabel('N칰mero de 칠xitos')
plt.ylabel('Frecuencia')
plt.xticks(np.arange(0, n+1, 1))
plt.show()
```

### 2.1.5 Variable Aleatoria Poisson

La distribuci칩n de Poisson es una distribuci칩n de probabilidad discreta que se utiliza para calcular la probabilidad de que ocurra un determinado n칰mero de eventos raros durante un intervalo dado (puede ser de tiempo, longitud, 치rea, etc.).

Esta variable aleatoria toma los valores sobre el conjunto $ \{0, 1, 2, \dots \} $ y tiene un par치metro $ \lambda > 0 $ , el cual representa el n칰mero de veces que se espera que ocurra un evento durante el intervalo dado.

La funci칩n de masa de probabilidades de una variable aleatoria Poisson, $ X $ , est치 dada por:

$$
f_X(k) =
\begin{cases}
e^{-\lambda} \frac{\lambda^k}{k!} & \quad \text{si } k = 0, 1, 2, \dots \\
0 & \quad \text{en otro caso}
\end{cases}
$$

En efecto, es una funci칩n de masa de probabilidades debido a que:

$$
\sum_{k=0}^{\infty} e^{-\lambda} \frac{\lambda^k}{k!} = e^{-\lambda} \sum_{k=0}^{\infty} \frac{\lambda^k}{k!}
$$

$$
= e^{-\lambda} \underbrace{\left(1 + \lambda + \frac{\lambda^2}{2!} + \frac{\lambda^3}{3!} + \cdots \right)}_{\text{desarrollo de Taylor de } e^\lambda}
$$

$$
= e^{-\lambda} e^\lambda = 1
$$

#### **Esperanza y Varianza**

- La **esperanza** de $ X $ es:

$$
\mathbb{E}[X] = \lambda
$$

- La **varianza** de $ X $ es:

$$
\text{Var}(X) = \lambda
$$

#### **Ejemplo**

Supongamos que el n칰mero de autos que pasan por una caseta de peaje por hora sigue una distribuci칩n Poisson con par치metro $ \lambda = 5 $ ,  es decir, se espera que pasen en promedio 5 autos por hora.

La probabilidad de que pasen exactamente 3 autos en una hora es:

$$
\mathbb{P}(X = 3) = e^{-5} \frac{5^3}{3!} = 0.1404
$$

La probabilidad de que pasen m치s de 4 autos (es decir, \( X > 4 \)) es:

$$
\mathbb{P}(X > 4) = 1 - \mathbb{P}(X \leq 4) = 1 - \left( \mathbb{P}(X=0) + \mathbb{P}(X=1) + \mathbb{P}(X=2) + \mathbb{P}(X=3) + \mathbb{P}(X=4) \right)
$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metro lambda de la distribuci칩n Poisson
lambda_param = 5

# N칰mero de experimentos
n_experiments = 1000

# Simulaci칩n de la variable aleatoria Poisson
poisson_samples = np.random.poisson(lambda_param, n_experiments)

# C치lculo de la esperanza y varianza simuladas
mean = np.mean(poisson_samples)
variance = np.var(poisson_samples)

print(f"Esperanza simulada: {mean}")
print(f"Varianza simulada: {variance}")

# Visualizaci칩n de los resultados
plt.hist(poisson_samples, bins=np.arange(min(poisson_samples), max(poisson_samples) + 1), edgecolor='black', alpha=0.7)
plt.title('Distribuci칩n de la variable aleatoria Poisson')
plt.xlabel('N칰mero de autos')
plt.ylabel('Frecuencia')
plt.xticks(np.arange(min(poisson_samples), max(poisson_samples) + 1, 1))
plt.show()
```

## Variable Aleatoria Geom칠trica

La variable aleatoria geom칠trica modela el n칰mero de **ensayos hasta obtener el primer 칠xito**, donde cada ensayo es independiente y tiene probabilidad de 칠xito $0 < p < 1$.

### Definici칩n

Si $X \sim \text{Geo}(p)$, su funci칩n de masa de probabilidades es:

$$
f_X(k) =
\begin{cases}
p(1-p)^{k-1} & \text{si } k = 1, 2, \dots \\
0 & \text{en otro caso}
\end{cases}
$$

Esto representa el n칰mero de **ensayos necesarios** para obtener el primer 칠xito.

### Alternativa

Si se desea modelar el **n칰mero de fracasos antes del primer 칠xito**, entonces:

$$
f_X(k) =
\begin{cases}
p(1-p)^k & \text{si } k = 0, 1, 2, \dots \\
0 & \text{en otro caso}
\end{cases}
$$

### Esperanza y Varianza

- $\mathbb{E}[X] = \frac{1}{p}$
- $\text{Var}(X) = \frac{1 - p}{p^2}$

### Ejemplo

La probabilidad de que un cliente compre un seguro en una llamada es $p = 0.2$. La probabilidad de lograr la primera venta en la **tercera llamada** es:

$$
\mathbb{P}(X = 3) = 0.2 \cdot (1 - 0.2)^2 = 0.128
$$

## Variable Aleatoria Binomial Negativa

Modela el n칰mero de **ensayos necesarios para obtener $r$ 칠xitos** en ensayos independientes con probabilidad de 칠xito $0 < p < 1$.

### Definici칩n

Si $X \sim \text{BN}(r, p)$, su funci칩n de masa de probabilidades es:

$$
f_X(n) =
\begin{cases}
\binom{n-1}{r-1} p^r (1 - p)^{n - r} & \text{si } n = r, r+1, \dots \\
0 & \text{en otro caso}
\end{cases}
$$

### Esperanza y Varianza

- $\mathbb{E}[X] = \frac{r}{p}$
- $\text{Var}(X) = \frac{r(1 - p)}{p^2}$

### Ejemplo

Se lanza una moneda con probabilidad de cara $p = 0.3$ hasta obtener 4 caras. La probabilidad de obtener la cuarta cara **en el s칠ptimo lanzamiento** es:

$$
\mathbb{P}(X = 7) = \binom{6}{3} (0.3)^4 (0.7)^3 = 0.0595
$$

## Variable Aleatoria Hipergeom칠trica

Se utiliza cuando se seleccionan elementos **sin reemplazo** de una poblaci칩n finita.

### Definici칩n

Sean $N$ elementos en total, de los cuales $m$ son considerados "칠xito", y se extraen $n$ elementos sin reemplazo. Si $X \sim \text{Hiper}(n, N, m)$, su funci칩n de masa de probabilidades es:

$$
f_X(i) = \frac{\binom{m}{i} \binom{N - m}{n - i}}{\binom{N}{n}} \quad \text{para } i = 0, 1, \dots, n
$$

### Esperanza y Varianza

- $\mathbb{E}[X] = \frac{n m}{N}$
- $\text{Var}(X) = \frac{n m}{N} \left(1 - \frac{m}{N}\right) \left( \frac{N - n}{N - 1} \right)$

### Aproximaci칩n

Si $m, N \gg n$, con $p = \frac{m}{N}$, entonces:

$$
\mathbb{P}(X = i) \approx \binom{n}{i} p^i (1 - p)^{n - i}
$$

### Ejemplo

Una urna contiene 10 bolas, 4 rojas y 6 verdes. Si se extraen 3 bolas sin reemplazo, la probabilidad de obtener exactamente 2 bolas rojas es:

$$
\mathbb{P}(X = 2) = \frac{\binom{4}{2} \binom{6}{1}}{\binom{10}{3}} = \frac{36}{120} = 0.3
$$



## 游늳 2.2 Variables Aleatorias Continuas
Pueden tomar cualquier valor dentro de un intervalo. Por ejemplo, la estatura de una persona:

- Se describe mediante una funci칩n de densidad $ f(x) $
- $ P(a \leq X \leq b) = \int_a^b f(x) \, dx $

F칩rmulas:
- Valor esperado: $ \mathbb{E}[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx $
- Varianza: $ \text{Var}(X) = \int_{-\infty}^{\infty} (x - \mathbb{E}[X])^2 \cdot f(x) \, dx $


## 2.2.1 Distribuci칩n Uniforme continua

Sea $X$ una variable aleatoria continua que se distribuye de forma uniforme en el intervalo $[a, b]$, donde $a < b$. Esto significa que todos los valores dentro del intervalo son igualmente probables, y fuera de 칠l, la probabilidad es cero.

Decimos que $X$ tiene una distribuci칩n uniforme continua si su funci칩n de densidad de probabilidad (f.d.p.) est치 dada por:

$$
f_X(x) =
\begin{cases}
\frac{1}{b - a}, & \text{si } x \in [a, b] \\
0, & \text{en otro caso}
\end{cases}
$$

Esta funci칩n cumple con las propiedades b치sicas de una funci칩n de densidad:

- $f_X(x) \geq 0$ para todo $x$.
- La integral sobre todo el espacio es 1:

$$
\int_{-\infty}^{\infty} f_X(x) \, dx = \int_a^b \frac{1}{b - a} \, dx = \frac{1}{b - a}(b - a) = 1
$$

### Esperanza y varianza

La esperanza y la varianza de una variable aleatoria $X \sim \text{Uniforme}(a, b)$ est치n dadas por:

- **Esperanza:**

$$
\mathbb{E}[X] = \frac{a + b}{2}
$$

- **Varianza:**

$$
\text{Var}(X) = \frac{(b - a)^2}{12}
$$

### Funci칩n de distribuci칩n acumulada (F.D.A.)

La funci칩n de distribuci칩n acumulada $F_X(x)$ est치 dada por:

$$
F_X(x) =
\begin{cases}
0, & \text{si } x < a \\
\frac{x - a}{b - a}, & \text{si } a \leq x \leq b \\
1, & \text{si } x > b
\end{cases}
$$

---

###  Ejemplo

Supongamos que el tiempo de espera (en minutos) para que pase un autob칰s est치 uniformemente distribuido entre 5 y 15 minutos. Entonces, si $X$ representa este tiempo de espera, tenemos que $X \sim \text{Uniforme}(5, 15)$.

- $f_X(x) = \frac{1}{15 - 5} = \frac{1}{10}$ para $x \in [5, 15]$.
- $\mathbb{E}[X] = \frac{5 + 15}{2} = 10$ minutos.
- $\text{Var}(X) = \frac{(15 - 5)^2}{12} = \frac{100}{12} \approx 8.33$

Calcular la probabilidad de que el autob칰s llegue entre los 8 y los 12 minutos:

$$
P(8 \leq X \leq 12) = \int_8^{12} \frac{1}{10} \, dx = \frac{12 - 8}{10} = 0.4
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Par치metros de la distribuci칩n
a = 5  # l칤mite inferior
b = 15  # l칤mite superior

# Creamos una distribuci칩n uniforme en [a, b]
dist = uniform(loc=a, scale=b - a)

# Valores para graficar
x = np.linspace(a - 2, b + 2, 1000)
y = dist.pdf(x)

# Graficamos la funci칩n de densidad
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='f.d.p. Uniforme[5, 15]', color='blue')
plt.fill_between(x, y, where=(x >= 8) & (x <= 12), color='skyblue', alpha=0.5,
                 label='츼rea entre 8 y 12')
plt.title('Funci칩n de densidad de una Uniforme continua')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# C치lculo de probabilidades
prob = dist.cdf(12) - dist.cdf(8)
print(f"La probabilidad de que X est칠 entre 8 y 12 es: {prob:.2f}")
```

## 2.2.2 Distribuci칩n Exponencial

Una variable aleatoria continua $X$ sigue una **Distribuci칩n Exponencial** con par치metro $\theta > 0$ si su funci칩n de densidad est치 dada por:

$$
f_X(x) = \begin{cases}
\theta e^{-\theta x} & \text{si } x \ge 0 \\
0 & \text{en otro caso}
\end{cases}
$$

Se denota $X \sim \text{Exp}(\theta)$. Esta distribuci칩n se utiliza frecuentemente para modelar **el tiempo entre eventos** en un proceso de Poisson.

La **funci칩n de distribuci칩n acumulada (FDA)** es:

$$
F_X(x) = \begin{cases}
1 - e^{-\theta x} & \text{si } x \ge 0 \\
0 & \text{en otro caso}
\end{cases}
$$

Las siguientes son propiedades importantes:
- Esperanza: $\mathbb{E}[X] = \frac{1}{\theta}$
- Varianza: $\text{Var}(X) = \frac{1}{\theta^2}$


```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metro theta
theta = 0.5

# Valores de x
x = np.linspace(0, 15, 1000)

# FDP y FDA
fdp = theta * np.exp(-theta * x)
fda = 1 - np.exp(-theta * x)

# Gr치fica de la FDP
plt.figure(figsize=(8, 4))
plt.plot(x, fdp, label='FDP', color='purple')
plt.title("Funci칩n de Densidad de Probabilidad (Exponencial)")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.grid()
plt.legend()
plt.show()

# Gr치fica de la FDA
plt.figure(figsize=(8, 4))
plt.plot(x, fda, label='FDA', color='green')
plt.title("Funci칩n de Distribuci칩n Acumulada (Exponencial)")
plt.xlabel("x")
plt.ylabel("Probabilidad acumulada")
plt.grid()
plt.legend()
plt.show()
```

### Ejemplo: Probabilidad entre dos valores

Supongamos $X \sim \text{Exp}(0.5)$. 쮺u치l es la probabilidad de que $X$ est칠 entre 2 y 5?

$$
\mathbb{P}(2 \le X \le 5) = F_X(5) - F_X(2) = \left(1 - e^{-0.5 \cdot 5}\right) - \left(1 - e^{-0.5 \cdot 2}\right)
$$


```python
# C치lculo usando FDA
from math import exp

p_2_5 = (1 - exp(-0.5 * 5)) - (1 - exp(-0.5 * 2))
print(f"P(2 <= X <= 5) = {p_2_5:.4f}")
```

### Aplicaci칩n: estimador de m치xima verosimilitud (MLE)

Supongamos que tenemos 5 observaciones independientes de una variable $X \sim \text{Exp}(\theta)$, y el total observado es 14. La funci칩n de verosimilitud est치 dada por:

$$
L(\theta) = \theta^5 e^{-14 \theta}
$$

Entonces, la log-verosimilitud es:

$$
\log L(\theta) = 5 \log \theta - 14\theta
$$

A continuaci칩n derivamos esta funci칩n y resolvemos para encontrar el valor de $\theta$ que maximiza la verosimilitud.


```python
import sympy as sp

# Definir la variable simb칩lica para theta
theta = sp.Symbol('theta', positive=True)

# Log-verosimilitud
log_likelihood = 5 * sp.log(theta) - 14 * theta

# Derivar y resolver
log_likelihood_derivative = sp.diff(log_likelihood, theta)
theta_mle = sp.solve(log_likelihood_derivative, theta)

# Resultado
print(f"El MLE para theta es: {theta_mle[0]}")
```

## 2.2.3 Distribuci칩n Normal

Una variable aleatoria continua $X$ sigue una **Distribuci칩n Normal** con media $\mu \in \mathbb{R}$ y desviaci칩n est치ndar $\sigma > 0$ si su funci칩n de densidad est치 dada por:

$$
f_X(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left( -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 \right), \quad x \in \mathbb{R}
$$

Se denota $X \sim \mathcal{N}(\mu, \sigma^2)$. Esta distribuci칩n es fundamental en estad칤stica y probabilidad debido al **Teorema Central del L칤mite** y su uso para modelar fen칩menos naturales y sociales.

### Propiedades importantes
- Sim칠trica respecto a $\mu$.
- Esperanza: $\mathbb{E}[X] = \mu$
- Varianza: $\text{Var}(X) = \sigma^2$
- La curva tiene forma de campana, conocida como **curva gaussiana**.

Cuando $\mu = 0$ y $\sigma = 1$, se dice que $X$ tiene una **Distribuci칩n Normal Est치ndar**, denotada $X \sim \mathcal{N}(0,1)$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par치metros de la distribuci칩n normal
mu = 5
sigma = 2

# Valores de x
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Funci칩n de densidad
fdp = norm.pdf(x, mu, sigma)

# Graficar la FDP
plt.figure(figsize=(8, 4))
plt.plot(x, fdp, label=f"$\mathcal{{N}}({mu}, {sigma**2})$", color='darkred')
plt.title("Funci칩n de Densidad de Probabilidad (Normal)")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.grid()
plt.legend()
plt.show()
```

### C치lculo de probabilidades

Supongamos que $X \sim \mathcal{N}(5, 4)$, es decir, $\mu = 5$ y $\sigma = 2$.

#### 쮺u치l es la probabilidad de que $X$ est칠 entre 3 y 7?
Esto se calcula como:

$$
\mathbb{P}(3 \le X \le 7) = F_X(7) - F_X(3)
$$

donde $F_X(x)$ es la funci칩n de distribuci칩n acumulativa.


```python
# C치lculo de la probabilidad P(3 곣 X 곣 7)
p_3_7 = norm.cdf(7, mu, sigma) - norm.cdf(3, mu, sigma)
print(f"P(3 곣 X 곣 7) = {p_3_7:.4f}")
```

### Estandarizaci칩n

Si $X \sim \mathcal{N}(\mu, \sigma^2)$, entonces la variable

$$
Z = \frac{X - \mu}{\sigma}
$$

sigue una distribuci칩n normal est치ndar: $Z \sim \mathcal{N}(0, 1)$. Esto permite usar las tablas de la normal est치ndar para calcular cualquier probabilidad.


## 2.2.4 Distribuci칩n Gamma

Una variable aleatoria continua $X$ tiene una **Distribuci칩n Gamma** con par치metros $\alpha > 0$ (forma) y $\theta > 0$ (escala) si su funci칩n de densidad est치 dada por:

$$
f_X(x) = \frac{1}{\Gamma(\alpha)\theta^\alpha} x^{\alpha - 1} e^{-x/\theta}, \quad x > 0
$$

donde $\Gamma(\alpha)$ es la funci칩n gamma, una generalizaci칩n del factorial: $\Gamma(n) = (n-1)!$ para $n \in \mathbb{N}$.

Notaci칩n: $X \sim \text{Gamma}(\alpha, \theta)$

**Propiedades**:
- Esperanza: $\mathbb{E}[X] = \alpha \theta$
- Varianza: $\text{Var}(X) = \alpha \theta^2$

La distribuci칩n exponencial es un caso particular de la Gamma con $\alpha = 1$.

## 2.2.5 Distribuci칩n Beta

Una variable aleatoria continua $X$ sigue una **Distribuci칩n Beta** con par치metros $\alpha, \beta > 0$ si su funci칩n de densidad est치 dada por:

$$
f_X(x) = \frac{1}{B(\alpha, \beta)} x^{\alpha - 1}(1 - x)^{\beta - 1}, \quad 0 < x < 1
$$

donde $B(\alpha, \beta)$ es la funci칩n beta, que normaliza la densidad y se define como:

$$
B(\alpha, \beta) = \int_0^1 t^{\alpha - 1}(1 - t)^{\beta - 1} dt = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

Notaci칩n: $X \sim \text{Beta}(\alpha, \beta)$

**Propiedades**:
- Esperanza: $\mathbb{E}[X] = \frac{\alpha}{\alpha + \beta}$
- Varianza: $\text{Var}(X) = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$

La Beta se usa com칰nmente en inferencia bayesiana y para modelar proporciones o probabilidades.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, beta

x_gamma = np.linspace(0, 20, 1000)
x_beta = np.linspace(0, 1, 1000)

# Par치metros para Gamma: (forma, escala)
gamma_params = [(1, 2), (2, 2), (5, 1)]

plt.figure(figsize=(12, 4))
for a, scale in gamma_params:
    plt.plot(x_gamma, gamma.pdf(x_gamma, a, scale=scale), label=fr'$\alpha={a}, \theta={scale}$')
plt.title("Distribuci칩n Gamma")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()

# Par치metros para Beta: (alpha, beta)
beta_params = [(0.5, 0.5), (2, 5), (5, 1)]

plt.figure(figsize=(12, 4))
for a, b in beta_params:
    plt.plot(x_beta, beta.pdf(x_beta, a, b), label=fr'$\alpha={a}, \beta={b}$')
plt.title("Distribuci칩n Beta")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```

# 2.3 Simulaci칩n de Variables Aleatorias

La simulaci칩n de variables aleatorias es un proceso mediante el cual generamos muestras aleatorias de una variable con una distribuci칩n dada. Este proceso es muy 칰til en situaciones donde obtener soluciones anal칤ticas exactas no es posible o cuando se requiere realizar experimentos para verificar teor칤as. Existen diversos m칠todos para realizar simulaciones, y algunos de los m치s comunes son los siguientes:

## 2.3.1 M칠todos de Simulaci칩n

### 1. M칠todo de la Transformada Inversa

El **m칠todo de la transformada inversa** es uno de los m치s comunes y sencillos. Este m칠todo se utiliza cuando conocemos la funci칩n de distribuci칩n acumulativa (FDA) de una variable aleatoria $X$ y podemos invertirla. El procedimiento es el siguiente:

1. **Generar un n칰mero aleatorio $u$**: Generamos un n칰mero aleatorio $u \sim U(0,1)$.
2. **Aplicar la inversa de la FDA**: Si conocemos la FDA de la variable aleatoria $X$, denotada por $F_X(x)$, entonces la variable aleatoria generada es:
   $$ X = F_X^{-1}(u) $$

#### Ejemplo con la Distribuci칩n Exponencial

Supongamos que queremos generar una variable aleatoria $X \sim \text{Exp}(\lambda)$, cuya FDA es:
$$ F_X(x) = 1 - e^{-\lambda x} $$

La inversa de esta FDA es:
$$ F_X^{-1}(u) = -\frac{1}{\lambda} \ln(1 - u) $$

Dado que $u \sim U(0, 1)$, podemos generar muestras de $X$ como:
$$ X = -\frac{1}{\lambda} \ln(u) $$

### 2. M칠todo de Aceptaci칩n-Rechazo

El **m칠todo de aceptaci칩n-rechazo** es m치s general y no requiere la inversi칩n de la FDA. En lugar de eso, se utiliza una funci칩n de densidad de probabilidad (FDP) que se pueda comparar f치cilmente con una distribuci칩n conocida. El algoritmo es el siguiente:

1. **Generar un valor $y$ de una distribuci칩n candidata**: Seleccionamos una distribuci칩n candidata f치cil de simular, de la cual generamos un valor $y$.
2. **Generar un valor de aceptaci칩n $u$**: Generamos un n칰mero aleatorio $u \sim U(0,1)$.
3. **Aceptar o rechazar**: Si $u \leq \frac{f_X(y)}{M f_Y(y)}$, donde $f_X(y)$ es la FDP de la distribuci칩n objetivo y $f_Y(y)$ es la FDP de la distribuci칩n candidata, aceptamos $y$ como una muestra de $X$. Si no, rechazamos $y$ y repetimos el proceso.

Este m칠todo es 칰til cuando la funci칩n de densidad objetivo no es f치cilmente invertible.

### 3. M칠todo de Monte Carlo

El **m칠todo de Monte Carlo** no est치 necesariamente relacionado con la generaci칩n directa de variables aleatorias, sino con la estimaci칩n de funciones matem치ticas mediante simulaciones. Se utiliza cuando no es posible obtener una soluci칩n cerrada de un problema, pero se puede estimar mediante la simulaci칩n.

El proceso b치sico es:

1. **Generar muestras aleatorias** de variables con distribuciones conocidas.
2. **Realizar experimentos** o c치lculos con esas muestras para estimar el valor de una funci칩n o un par치metro.

Por ejemplo, se podr칤a utilizar el m칠todo de Monte Carlo para estimar la integral de una funci칩n o para aproximar el valor de una expectativa.

### 4. M칠todo de la Serie de Taylor (o de Aproximaci칩n Estoc치stica)

El **m칠todo de la serie de Taylor** se usa cuando se tiene una funci칩n de distribuci칩n dif칤cil de simular. En este m칠todo, se emplea una expansi칩n de Taylor de una funci칩n aleatoria para aproximar su valor esperado o su distribuci칩n. Es 칰til para generar muestras de distribuciones que se relacionan con distribuciones conocidas, mediante la adici칩n de perturbaciones aleatorias.

Este m칠todo es m치s com칰n en aplicaciones avanzadas, como en la simulaci칩n de procesos estoc치sticos.

## Elecci칩n del M칠todo de Simulaci칩n

La elecci칩n de un m칠todo depende de varios factores:
- **Facilidad de simulaci칩n**: Algunas distribuciones son m치s f치ciles de simular que otras (por ejemplo, la distribuci칩n uniforme).
- **Requerimientos de precisi칩n**: Si se necesita una simulaci칩n precisa, se puede preferir el m칠todo de la transformada inversa, siempre que sea posible.
- **Disponibilidad de funciones**: Algunos lenguajes de programaci칩n tienen bibliotecas para generar distribuciones espec칤ficas de manera eficiente.

En resumen, los m칠todos de simulaci칩n permiten generar muestras aleatorias de diferentes distribuciones, facilitando el an치lisis de fen칩menos estoc치sticos en diversas aplicaciones.

## Ejemplo 1. Simulaci칩n de Monte Carlo para estimar la integral de una funci칩n

```python
import numpy as np

# Funci칩n a integrar
def f(x):
    return np.sin(x)  # Ejemplo con la funci칩n seno

# M칠todo de Monte Carlo para estimar la integral de f(x) desde a hasta b
def monte_carlo_integration(f, a, b, n):
    x = np.random.uniform(a, b, n)  # Generar n muestras uniformemente distribuidas entre a y b
    y = f(x)  # Evaluar la funci칩n en esas muestras
    return (b - a) * np.mean(y)  # Aproximaci칩n a la integral

# Par치metros
a, b = 0, np.pi  # Intervalo de integraci칩n
n = 100000  # N칰mero de muestras

# Estimaci칩n de la integral
integral_estimada = monte_carlo_integration(f, a, b, n)
print(f"Estimaci칩n de la integral de sin(x) entre 0 y pi: {integral_estimada:.4f}")
```

```python
import matplotlib.pyplot as plt

# Par치metros de la simulaci칩n
mean_demand = 50  # Media de la demanda diaria
std_demand = 10   # Desviaci칩n est치ndar de la demanda diaria
stock_level = 60  # Nivel de inventario inicial

# N칰mero de simulaciones
n_simulations = 1000

# Funci칩n para simular el comportamiento de la demanda
def simulate_inventory(n_simulations, mean_demand, std_demand, stock_level):
    days = np.random.normal(mean_demand, std_demand, n_simulations)  # Demanda diaria simulada
    shortages = np.sum(days > stock_level)  # N칰mero de d칤as con desabastecimiento
    return shortages / n_simulations  # Proporci칩n de d칤as con desabastecimiento

# Simulaci칩n
risk = simulate_inventory(n_simulations, mean_demand, std_demand, stock_level)
print(f"Riesgo de desabastecimiento (proporci칩n de d칤as): {risk:.4f}")

# Graficar la distribuci칩n de las demandas simuladas
plt.hist(np.random.normal(mean_demand, std_demand, n_simulations), bins=30, color='blue', alpha=0.7)
plt.title('Distribuci칩n Simulada de Demanda Diaria')
plt.xlabel('Demanda')
plt.ylabel('Frecuencia')
plt.show()
```

# 3.1 Funci칩n Generadora de Momentos (FGM)

### Definici칩n

La **Funci칩n Generadora de Momentos (FGM)** de una variable aleatoria \( X \) es una funci칩n matem치tica que se utiliza para obtener los momentos (es decir, la esperanza, la varianza, etc.) de la distribuci칩n de \( X \). Est치 definida como el valor esperado de la exponencial de \( tX \), es decir:

$$
M_X(t) = \mathbb{E}[e^{tX}]
$$

Donde:
- \( M_X(t) \) es la FGM de la variable aleatoria \( X \),
- \( t \) es un par치metro real, y
- \( \mathbb{E}[e^{tX}] \) es el valor esperado de la funci칩n \( e^{tX} \).

### Propiedades de la Funci칩n Generadora de Momentos

1. **Existencia**:
   La FGM de \( X \) existe en un intervalo abierto que contiene \( t = 0 \) (siempre que \( X \) tenga momentos finitos).

2. **Momento de orden \( n \)**:
   El \( n \)-칠simo momento de \( X \), \( \mathbb{E}[X^n] \), puede obtenerse a partir de la FGM derivando la funci칩n generadora de momentos \( n \) veces con respecto a \( t \) y evaluando en \( t = 0 \):
   $$
   \mathbb{E}[X^n] = \frac{d^n M_X(t)}{dt^n} \Big|_{t=0}
   $$

3. **Propiedad de la linealidad**:
   Si \( X \) y \( Y \) son variables aleatorias, la FGM de la suma \( X + Y \) es el producto de las FGMs de \( X \) y \( Y \), siempre que \( X \) y \( Y \) sean independientes:
   $$
   M_{X+Y}(t) = M_X(t) \cdot M_Y(t)
   $$

4. **Funci칩n generadora de momentos de variables transformadas**:
   Si \( Y = aX + b \), entonces la FGM de \( Y \) es:
   $$
   M_Y(t) = e^{bt} M_X(at)
   $$

5. **Determinar la distribuci칩n**:
   En algunos casos, la FGM es 칰nica para una distribuci칩n, lo que significa que la FGM puede ser utilizada para identificar completamente la distribuci칩n de una variable aleatoria.

### Aplicaciones de la Funci칩n Generadora de Momentos

- **C치lculo de momentos**: La FGM es 칰til para calcular momentos de una distribuci칩n sin necesidad de recurrir a la definici칩n directa de \( \mathbb{E}[X^n] \).
  
- **Teor칤a de colas**: En procesos estoc치sticos, como las colas de espera, la FGM se utiliza para modelar el comportamiento de las llegadas de clientes y las esperas en sistemas de servicio.
  
- **Distribuciones compuestas**: La FGM es 칰til para obtener la distribuci칩n de la suma de variables aleatorias, especialmente cuando estas variables son independientes.

- **C치lculo de la varianza y la covarianza**: La FGM es 칰til para obtener varianzas y covarianzas en distribuciones complejas o para obtener los momentos de distribuci칩n de una suma de variables aleatorias.



### Ejemplo: Visualizaci칩n de la FGM para una distribuci칩n exponencial

Consideremos una variable aleatoria $ X \sim \text{Exponencial}(\lambda) $ La FGM de $ X $ es:

$$
M_X(t) = \mathbb{E}[e^{tX}] = \frac{\lambda}{\lambda - t}, \quad \text{para } t < \lambda
$$

Esta funci칩n est치 definida solo para $ t < \lambda $ ya que fuera de ese rango la esperanza no converge.

A continuaci칩n, graficamos esta funci칩n para visualizar c칩mo se comporta.


```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metro de la distribuci칩n exponencial
lambd = 2  # lambda > 0

# Rango de t para el que M_X(t) est치 definido: t < lambda
t_values = np.linspace(-1, lambd - 0.01, 500)
mgf_values = lambd / (lambd - t_values)

plt.figure(figsize=(8, 5))
plt.plot(t_values, mgf_values, label=r'$M_X(t) = \frac{\lambda}{\lambda - t}$', color='purple')
plt.axvline(lambd, color='red', linestyle='--', label=r'L칤mite $t = \lambda$')
plt.title('Funci칩n Generadora de Momentos (FGM) - Distribuci칩n Exponencial')
plt.xlabel('t')
plt.ylabel(r'$M_X(t)$')
plt.ylim(0, 10)
plt.grid(True)
plt.legend()
plt.show()
```

## 3.2 Funci칩n Caracter칤stica (FC)

### Definici칩n

La **funci칩n caracter칤stica** de una variable aleatoria $ X $ es la funci칩n compleja definida por:

$$
\varphi_X(t) = \mathbb{E}[e^{itX}], \quad t \in \mathbb{R}
$$

donde $ i $ es la unidad imaginaria, es decir, $ i^2 = -1 $

La funci칩n caracter칤stica **siempre existe** (a diferencia de la FGM) y determina completamente la distribuci칩n de una variable aleatoria.

---

### Aplicaciones y ventajas

- Permite derivar momentos de una variable aleatoria (cuando existen).
- 칔til para estudiar **convergencia en distribuci칩n** (teorema central del l칤mite).
- La funci칩n caracter칤stica de la **suma de variables independientes** es el producto de sus funciones caracter칤sticas.
- Permite derivar distribuciones l칤mite en procesos estoc치sticos.
- A diferencia de la FGM, **existe para todas las variables aleatorias**.

---

### Ejemplo: Funci칩n caracter칤stica de la distribuci칩n normal

Sea $ X \sim \mathcal{N}(\mu, \sigma^2) $ su funci칩n caracter칤stica es:

$$
\varphi_X(t) = e^{it\mu - \frac{1}{2} \sigma^2 t^2}
$$

Vamos a graficar la parte real e imaginaria de $ \varphi_X(t) $ para distintos valores de $ t $



```python
import numpy as np
import matplotlib.pyplot as plt

# Par치metros de la normal
mu = 0.5
sigma = 1

# Funci칩n caracter칤stica de la normal
def phi_X(t, mu=mu, sigma=sigma):
    return np.exp(1j * t * mu - 0.5 * (sigma**2) * t**2)

# Valores de t
t_vals = np.linspace(-10, 10, 500)
phi_vals = phi_X(t_vals)

# Graficar parte real e imaginaria
plt.figure(figsize=(10, 5))
plt.plot(t_vals, np.real(phi_vals), label='Parte real', color='blue')
plt.plot(t_vals, np.imag(phi_vals), label='Parte imaginaria', color='orange')
plt.title('Funci칩n Caracter칤stica de una Variable Aleatoria Normal')
plt.xlabel('t')
plt.ylabel(r'$\varphi_X(t)$')
plt.axhline(0, color='black', linestyle='--', alpha=0.6)
plt.legend()
plt.grid(True)
plt.show()
```

## 4. Teoremas Fundamentales

### 4.1 Ley de los Grandes N칰meros (LGN)

La **Ley de los Grandes N칰meros** es un resultado esencial de la teor칤a de la probabilidad. Describe el comportamiento del promedio de una secuencia de variables aleatorias conforme el tama침o de la muestra crece. Tiene dos versiones principales: **forma d칠bil** y **forma fuerte**.

---

###  Enunciado (Forma D칠bil)

Sea ${X_1, X_2, \dots, X_n}$ una sucesi칩n de variables aleatorias i.i.d. (independientes e id칠nticamente distribuidas) con esperanza finita $\mu = \mathbb{E}[X_i]$. Entonces, el promedio muestral:

$$
\bar{X}_n = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

converge en probabilidad al valor esperado $\mu$, es decir:

$$
\bar{X}_n \xrightarrow{p} \mu \quad \text{cuando} \quad n \to \infty
$$

---

###  Enunciado (Forma Fuerte)

Bajo las mismas condiciones, si adem치s $\mathbb{E}[|X_i|] < \infty$, entonces:

$$
\bar{X}_n \xrightarrow{c.s.} \mu \quad \text{cuando} \quad n \to \infty
$$

Esto significa que $\bar{X}_n$ converge **casi seguramente** a $\mu$ (con probabilidad 1), lo cual es una forma m치s fuerte de convergencia.

---

###  Implicaciones

- Permite justificar el uso del promedio muestral $\bar{X}_n$ como estimador del valor esperado poblacional $\mu$.
- Fundamenta el uso de simulaciones en estad칤stica aplicada.
- Garantiza que con suficientes observaciones, el comportamiento aleatorio se estabiliza.
- Es la base de la inferencia estad칤stica cl치sica.


```python
import numpy as np
import matplotlib.pyplot as plt

# Fijar la semilla para reproducibilidad
np.random.seed(0)

# Par치metros
n = 10000  # n칰mero de observaciones
mu = 5     # valor esperado verdadero
sigma = 2  # desviaci칩n est치ndar

# Generar variables aleatorias normales
X = np.random.normal(loc=mu, scale=sigma, size=n)

# Promedios muestrales acumulados
media_muestral = np.cumsum(X) / np.arange(1, n + 1)

# Gr치fica
plt.figure(figsize=(10, 5))
plt.plot(media_muestral, label='Media muestral acumulada')
plt.axhline(y=mu, color='red', linestyle='--', label=r'$\mu = 5$')
plt.title("Ley de los Grandes N칰meros (Simulaci칩n)")
plt.xlabel("N칰mero de observaciones")
plt.ylabel("Media muestral")
plt.grid(True)
plt.legend()
plt.show()
```

## 4.2 Teorema Central del L칤mite (TCL)

El **Teorema Central del L칤mite** es uno de los resultados m치s importantes de la estad칤stica y la probabilidad. Establece que, bajo ciertas condiciones, el comportamiento de la suma (o promedio) de variables aleatorias puede aproximarse por una distribuci칩n normal.

---

###  Enunciado

Sea ${X_1, X_2, \dots, X_n}$ una secuencia de variables aleatorias i.i.d. con valor esperado $\mu = \mathbb{E}[X_i]$ y varianza finita $\sigma^2 = \text{Var}(X_i)$. Entonces, al aumentar el tama침o de la muestra $n$, la variable aleatoria estandarizada:

$$
Z_n = \frac{\bar{X}_n - \mu}{\sigma / \sqrt{n}} = \frac{1}{\sqrt{n}} \sum_{i=1}^{n} \frac{X_i - \mu}{\sigma}
$$

converge en distribuci칩n a una normal est치ndar:

$$
Z_n \xrightarrow{d} N(0, 1)
$$

---

###  Aplicaci칩n pr치ctica

El **TCL** permite usar la distribuci칩n normal como aproximaci칩n en problemas donde se trabaja con sumas o promedios de muchas variables aleatorias, incluso si las variables originales no son normales.

**Aplicaciones clave:**

- Intervalos de confianza y pruebas de hip칩tesis.
- Simulaci칩n estad칤stica.
- Control de calidad.
- An치lisis de grandes bases de datos.

---

###  Relaci칩n con la distribuci칩n normal

El TCL explica por qu칠 la normal aparece con tanta frecuencia en estad칤sticas muestrales. Aunque la poblaci칩n no sea normal, el promedio muestral se aproxima a una distribuci칩n normal cuando $n$ es grande.

Esto permite usar la normal como aproximaci칩n para otras distribuciones, por ejemplo:

- Binomial con $n$ grande.
- Poisson con $\lambda$ grande.
- Cualquier distribuci칩n con media y varianza finitas.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Fijar semilla para reproducibilidad
np.random.seed(42)

# Par치metros
n = 30          # tama침o de muestra
repeticiones = 10000
lambda_pois = 4  # par치metro de una distribuci칩n Poisson

# Simular muchas muestras de promedio de Poisson
medias = [np.mean(np.random.poisson(lam=lambda_pois, size=n)) for _ in range(repeticiones)]

# Gr치fica del histograma de los promedios
plt.figure(figsize=(10,5))
plt.hist(medias, bins=50, density=True, alpha=0.6, color='skyblue', label='Distribuci칩n emp칤rica')

# Superponer la normal te칩rica (TCL)
mu = lambda_pois
sigma = np.sqrt(lambda_pois / n)
x = np.linspace(min(medias), max(medias), 200)
plt.plot(x, norm.pdf(x, loc=mu, scale=sigma), 'r--', label='Normal aprox. (TCL)')

plt.title("Teorema Central del L칤mite con muestras Poisson")
plt.xlabel("Promedio muestral")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```
