## *Probabilidad Laplaciana*

### 1.1.1 Definici√≥n de la Probabilidad Laplaciana

La **probabilidad Laplaciana**, tambi√©n llamada **probabilidad cl√°sica**, se basa en el principio de **equiprobabilidad**. Se define matem√°ticamente como:

$$P(A) = \frac{\text{Casos favorables}}{\text{Casos posibles}}
$$

donde:

- \( P(A) \) representa la probabilidad del evento \( A \).
- **"Casos favorables"** son los resultados dentro del espacio muestral que satisfacen \( A \).
- **"Casos posibles"** es el n√∫mero total de resultados en el espacio muestral \( \Omega \).

Este modelo de probabilidad es v√°lido √∫nicamente cuando **todos los resultados son igualmente probables**. Es decir, si se lanza un dado justo, **cada cara tiene la misma probabilidad de salir**.

### 1.1.2  Teorema Central del Espacio de Equiprobabilidad
El Teorema Central de la Probabilidad Laplaciana establece que si un espacio muestral Œ© tiene ùëÅ elementos equiprobables, y un evento ùê¥ tiene ùëÄ de esos elementos, entonces la probabilidad de que ocurra ùê¥ est√° dada por:

$$ùëÉ (ùê¥) = \frac{‚à£ùê¥‚à£}{‚à£Œ©‚à£}$$
donde:

‚à£ùê¥‚à£ es el n√∫mero de elementos que componen el evento ùê¥.

‚à£ Œ© ‚à£ es el n√∫mero de elementos del espacio muestral.

Este teorema es la base fundamental de la probabilidad cl√°sica y se usa en experimentos como lanzamiento de monedas, dados y cartas.

### 1.1.3 Demostraci√≥n del Teorema
Sabemos que en un experimento con equiprobabilidad, todos los casos posibles tienen igual probabilidad de ocurrir. Sea
Œ© el espacio muestral con ùëÅ resultados posibles, cada uno con probabilidad:

$$ùëÉ(ùúî_ùëñ)=\frac{1}{ùëÅ}, \hspace{0.5cm} ‚àÄ  ùúî_ùëñ ‚àà Œ©$$
Si el evento ùê¥ est√° compuesto por ùëÄ elementos, entonces la probabilidad de ocurrencia de ùê¥ es la suma de las probabilidades de sus elementos individuales:
$$ ùëÉ (ùê¥)=‚àë_{ùúî_ùëñ ‚ààùê¥}ùëÉ(ùúî_ùëñ)$$
Como cada ùëÉ(ùúî_ùëñ) es igual a $\frac{1}{ùëÅ}$, tenemos:
$$
ùëÉ(ùê¥)=‚àë_{ùúî_ùëñ ‚ààùê¥}\frac{1}{ùëÅ}$$
Dado que hay ùëÄ elementos en ùê¥:
$$
ùëÉ(ùê¥)=ùëÄ‚ãÖ\frac{1}{ùëÅ}=ùëÄ$$

### 1.2.1 Axiomas de Kolmogorov: Fundamento de la Probabilidad Moderna

La teor√≠a de la probabilidad, tal como la conocemos hoy, se basa en la **formalizaci√≥n matem√°tica** de **Andrei Kolmogorov** en 1933. √âl estableci√≥ tres axiomas fundamentales que definen cualquier medida de probabilidad sobre un espacio muestral. Estos axiomas son la base sobre la que se construyen conceptos m√°s avanzados como la probabilidad condicional, independencia y distribuciones probabil√≠sticas.

### 1.2.2 Axioma de No Negatividad

La probabilidad de cualquier evento $A$ en el espacio muestral $\Omega$  debe ser mayor o igual a cero:

$$
P(A) \geq 0, \quad \forall A \subseteq \Omega
$$

- **Importancia**: La probabilidad es una medida cuantitativa de incertidumbre. Nunca puede ser negativa, pues representa la proporci√≥n de ocurrencia de un evento.
  
- **Aplicaci√≥n**: Si lanzamos un dado justo, la probabilidad de obtener cualquier n√∫mero v√°lido est√° entre 0 y 1. La probabilidad de obtener un n√∫mero menor que 1 es 0 (evento imposible).


### 1.2.3 Axioma de Normalizaci√≥n

La probabilidad del espacio muestral completo debe ser igual a 1:

$$
P(\Omega) = 1
$$

- **Importancia**: Este axioma establece que alg√∫n resultado del espacio muestral debe ocurrir. Es decir, en cualquier experimento hay certeza de que al menos un resultado se manifiesta.
  
- **Aplicaci√≥n**: En un lanzamiento de una moneda, el espacio muestral es $ \Omega = \{ \text{cara}, \text{cruz} \} $, y sabemos que seguro obtendremos una de esas opciones, por lo que:

$$
P(\text{cara}) + P(\text{cruz}) = 1
$$


### 1.2.4 Axioma de Aditividad (para eventos mutuamente excluyentes)

Si dos eventos $ A $ y $ B $ no tienen resultados en com√∫n, entonces:

$$
P(A \cup B) = P(A) + P(B)
$$

- **Importancia**: Si dos eventos no pueden ocurrir simult√°neamente, la probabilidad de que ocurra uno u otro es simplemente la suma de sus probabilidades individuales.
  
- **Aplicaci√≥n**: Supongamos que sacamos una carta de un mazo est√°ndar de 52 cartas. La probabilidad de sacar un As es:

$$
P(A) = \frac{4}{52}
$$

La probabilidad de sacar una Reina es:

$$
P(B) = \frac{4}{52}
$$

Como no podemos sacar un As y una Reina a la vez, aplicamos la aditividad:

$$
P(A \cup B) = P(A) + P(B) = \frac{4}{52} + \frac{4}{52} = \frac{8}{52}
$$

```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
from itertools import product
from collections import Counter
```

```python
# Espacio muestral
espacio_muestral = [1, 2, 3, 4, 5, 6]

# Casos favorables: n√∫meros pares
favorables = [x for x in espacio_muestral if x % 2 == 0]

# Probabilidad cl√°sica
p_par = len(favorables) / len(espacio_muestral)
print(f"La probabilidad de obtener un n√∫mero par es: {p_par:.2f}")
```

```python
# Espacio muestral: todos los pares (d1, d2)
espacio = list(itertools.product(range(1, 7), repeat=2))
n_total = len(espacio)

# Evento: suma igual a 7
suma_frecuencias = [sum(p) for p in espacio]
frecuencias = Counter(suma_frecuencias)
n_favorables = frecuencias[7]

# Probabilidad cl√°sica
P = n_favorables / n_total
print(f"Probabilidad de suma 7: {P:.4f}")

# Crear histograma
plt.hist(suma_frecuencias, bins=range(2, 14), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.axvline(7, color='red', linestyle='--', label='Suma 7')
plt.title('Distribuci√≥n de Sumas de Dos Dados')
plt.xlabel('Suma')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar probabilidad en el gr√°fico
plt.text(7, max(plt.hist(suma_frecuencias, bins=range(2, 14), align='left', rwidth=0.8)[0]),
         f'P(7) = {P:.4f}', ha='center', va='bottom', color='red')

plt.show()
```

## Probabilidad Geom√©trica

La **probabilidad geom√©trica** se aplica cuando el espacio muestral es un conjunto continuo y la probabilidad de un evento se determina en funci√≥n de la "medida" del conjunto favorable en relaci√≥n al tama√±o total del espacio muestral. Es decir, la probabilidad de un evento es proporcional al tama√±o del conjunto favorable dentro del espacio muestral.

### F√≥rmula General

Sea $A$ un evento en un espacio muestral continuo $\Omega$, la probabilidad de $A$ se calcula como:

$$
P(A) = \frac{\text{medida}(A)}{\text{medida}(\Omega)}
$$

Donde:
- $A$ es el conjunto de eventos favorables,
- $\Omega$ es el espacio muestral total,
- $\text{medida}(A)$ representa el "tama√±o" del conjunto favorable, que puede ser una longitud, √°rea, volumen, etc.
- $\text{medida}(\Omega)$ es el "tama√±o" total del espacio muestral.


### Aplicaciones

Este tipo de probabilidad es √∫til en muchos campos, como la **geometr√≠a** y la **f√≠sica**, cuando se modelan fen√≥menos aleatorios en espacios continuos. Adem√°s, es especialmente relevante cuando trabajamos con **modelos de √°reas** o **vol√∫menes**.

```python
# Contexto:
# El problema de la aguja de Buffon consiste en estimar el valor de pi mediante la probabilidad de que
# una aguja de longitud L, lanzada aleatoriamente sobre l√≠neas paralelas separadas por una distancia d,
# cruce alguna de las l√≠neas. Cuantas m√°s veces lanzamos la aguja, mejor ser√° la estimaci√≥n de pi.
# Este c√≥digo simula el problema de la aguja de Buffon para estimar el valor de pi.
# Par√°metros

import numpy as np
import matplotlib.pyplot as plt

# Par√°metros del problema
L = 1.0  # Longitud de la aguja
d = 1.0  # Distancia entre las l√≠neas paralelas
num_trials = 100000  # N√∫mero total de lanzamientos
step = 1000  # N√∫mero de lanzamientos por paso para la estimaci√≥n de pi

# Listas para almacenar los valores de pi estimados y los lanzamientos
pi_estimates = []
trial_steps = []

# Contador para cu√°ntas agujas cruzan una l√≠nea
crossed_lines = 0

# Simulaci√≥n de lanzamientos de la aguja y estimaci√≥n de pi en pasos
for trial in range(1, num_trials + 1):
    # Posici√≥n del centro de la aguja
    center_position = np.random.uniform(0, d / 2)

    # √Ångulo aleatorio entre 0 y pi (√°ngulo en radianes)
    angle = np.random.uniform(0, np.pi)

    # Distancia desde el centro de la aguja hasta la l√≠nea m√°s cercana
    distance_to_nearest_line = L / 2 * np.sin(angle)

    # Verificamos si la aguja cruza una l√≠nea
    if center_position <= distance_to_nearest_line:
        crossed_lines += 1

    # Estimaci√≥n de pi en pasos
    if trial % step == 0 and crossed_lines > 0:
        pi_estimate = (2 * L * trial) / (crossed_lines * d)
        pi_estimates.append(pi_estimate)
        trial_steps.append(trial)

print(f"Estimaci√≥n de pi: {pi_estimate}")

# Plot de la estimaci√≥n de pi a lo largo de los lanzamientos
plt.figure(figsize=(10, 6))
plt.plot(trial_steps, pi_estimates, label="Estimaci√≥n de œÄ", color="blue")
plt.axhline(y=np.pi, color="red", linestyle="--", label="Valor real de œÄ")
plt.xlabel('N√∫mero de lanzamientos')
plt.ylabel('Estimaci√≥n de œÄ')
plt.title('Convergencia de la estimaci√≥n de œÄ con el problema de la aguja de Buffon')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Problema de la Zona de Intersecci√≥n de dos C√≠rculos: Con ayuda de la probabilidad geometrica y segmentos circulares,
#se calcular√° la zona compartida entre dos c√≠rculos en funci√≥n de sus radios y la distancia entre sus centros.
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros
r = 1  # Radio de los c√≠rculos
n = 22000  # N√∫mero de simulaciones
intersecciones = 0

# Funci√≥n para simular la posici√≥n de los c√≠rculos
def circulos_interseccion():
    # Posiciones aleatorias del centro de los c√≠rculos
    x1, y1 = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
    x2, y2 = np.random.uniform(-5, 5), np.random.uniform(-5, 5)

    # Calcular la distancia entre los centros
    distancia_centros = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    # Si la distancia entre los centros es menor o igual a 2*r, los c√≠rculos se intersectan
    return distancia_centros <= 2 * r

# Simulaci√≥n
for _ in range(n):
    if circulos_interseccion():
        intersecciones += 1

# Estimaci√≥n de la probabilidad
probabilidad = intersecciones / n
print(f"Probabilidad de intersecci√≥n de los c√≠rculos: {probabilidad:.4f}")

# Visualizaci√≥n de una simulaci√≥n aleatoria
x1, y1 = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
x2, y2 = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
fig, ax = plt.subplots()

# C√≠rculo 1
circle1 = plt.Circle((x1, y1), r, color='blue', alpha=0.5)
ax.add_artist(circle1)

# C√≠rculo 2
circle2 = plt.Circle((x2, y2), r, color='red', alpha=0.5)
ax.add_artist(circle2)

# Configuraci√≥n de la gr√°fica
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_aspect('equal', 'box')
ax.set_title('Simulaci√≥n de Intersecci√≥n de C√≠rculos')
plt.show()
```

## **Probabilidad Frecuentista**

La **probabilidad frecuentista** se basa en la observaci√≥n de la frecuencia relativa de los eventos en experimentos repetidos. Seg√∫n este enfoque, la probabilidad de un evento se define como el **l√≠mite** de la frecuencia relativa de su ocurrencia conforme el n√∫mero de experimentos aumenta infinitamente.

Este enfoque es especialmente √∫til cuando no conocemos la probabilidad exacta de un evento, pero podemos **realizar experimentos** y **contabilizar** cu√°ntas veces ocurre el evento en cuesti√≥n.

### **F√≥rmula General**

Supongamos que repetimos un experimento aleatorio $n$ veces, y el evento $A$ ocurre $f(A)$ veces. Entonces, la **probabilidad frecuentista** de que ocurra el evento $A$ se calcula como:

$$
P(A) = \lim_{n \to \infty} \frac{f(A)}{n}
$$

Donde:
- $f(A)$ es el n√∫mero de veces que ocurre el evento $A$ en $n$ experimentos,
- $n$ es el n√∫mero total de experimentos.

A medida que $n$ se hace muy grande, la **frecuencia relativa** de la ocurrencia del evento $A$ se aproxima a su **probabilidad verdadera**.

### **Conceptos Clave**

1. **Experimentos Aleatorios**: Un **experimento aleatorio** es un proceso cuyo resultado no es predecible con certeza, pero se conoce el conjunto de todos los resultados posibles (espacio muestral).

2. **Frecuencia Relativa**: La frecuencia relativa de un evento es la **proporci√≥n de veces** que ocurre en relaci√≥n con el n√∫mero total de experimentos realizados. Se calcula como:

$$
\text{Frecuencia Relativa} = \frac{f(A)}{n}
$$

3. **Ley de los Grandes N√∫meros**: Seg√∫n esta ley, conforme el n√∫mero de experimentos aumenta, la frecuencia relativa de un evento se estabiliza y se aproxima a su probabilidad te√≥rica. Formalmente, se expresa como:

$$
\lim_{n \to \infty} \frac{f(A)}{n} = P(A)
$$

4. **Independencia de Experimentos**: En el enfoque frecuentista, los experimentos son **independientes**

```python
#Ejemplo del Volado

# Fijar la semilla para reproducibilidad
np.random.seed(42)

# Definir par√°metros
num_lanzamientos = 10_000
opciones = ['sol', 'aguila']

# Simulaci√≥n de lanzamientos
resultados = np.random.choice(opciones, size=num_lanzamientos)

# C√°lculo de frecuencia relativa acumulada
frecuencia_sol = np.cumsum(resultados == 'sol') / np.arange(1, num_lanzamientos + 1)

# Configuraci√≥n de la gr√°fica
plt.figure(figsize=(10, 5))
plt.plot(frecuencia_sol, label='Frecuencia relativa de "sol"', color='blue')
plt.axhline(0.5, color='red', linestyle='--', label='Valor te√≥rico: 0.5')

# Personalizaci√≥n de etiquetas y t√≠tulo
plt.title('Probabilidad Frecuentista: Convergencia de Frecuencia Relativa')
plt.xlabel('N√∫mero de lanzamientos')
plt.ylabel('Frecuencia relativa')
plt.legend()
plt.grid(True)

# Mostrar gr√°fica
plt.show()
```

```python
# Simulaci√≥n de un proceso de Poisson
np.random.seed(42)
n = 10000
lambda_poisson = 3  # Tasa de ocurrencias
resultados_poisson = np.random.poisson(lambda_poisson, n)

# Frecuencia relativa de los eventos
frecuencia_poisson = np.cumsum(resultados_poisson == 3) / np.arange(1, n + 1)

# Graficar
plt.figure(figsize=(10, 5))
plt.plot(frecuencia_poisson, label=f'Frecuencia relativa de ocurrencias de 3', color='green')
plt.axhline(1/np.exp(lambda_poisson), color='red', linestyle='--', label=f'Valor te√≥rico para {lambda_poisson}')
plt.title('Probabilidad Frecuentista: Proceso de Poisson')
plt.xlabel('N√∫mero de experimentos')
plt.ylabel('Frecuencia relativa')
plt.grid(True)
plt.legend()
plt.show()
```

## **Probabilidad Conjunta**

La **probabilidad conjunta** se refiere a la probabilidad de que ocurran dos (o m√°s) eventos al mismo tiempo. Es decir, es la probabilidad de que los eventos $A$ y $B$ ocurran simult√°neamente, y se den en conjunto. Se denota como $P(A \cap B)$, donde "$\cap$" indica la intersecci√≥n de los eventos.

### **F√≥rmula General**

La probabilidad conjunta de dos eventos $A$ y $B$ es:

$$
P(A \cap B) = P(A \mid B) \cdot P(B)
$$

Donde:
- $P(A \mid B)$ es la **probabilidad condicionada** de que ocurra el evento $A$, dado que ya ocurri√≥ el evento $B$.
- $P(B)$ es la probabilidad de que ocurra el evento $B$.

### **Condici√≥n de Independencia**

Si los eventos $A$ y $B$ son **independientes**, entonces la probabilidad conjunta se simplifica a:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

Esto significa que la ocurrencia de uno de los eventos no afecta la probabilidad de que ocurra el otro evento.

### **Aplicaciones**

La probabilidad conjunta se utiliza en una variedad de situaciones, como:
- **Problemas de urnas**: ¬øCu√°l es la probabilidad de sacar una bola roja y una bola azul de una urna?
- **Encuestas y estudios de mercado**: Calcular la probabilidad de que una persona tenga ciertas caracter√≠sticas simult√°neamente (por ejemplo, que sea hombre y fumador).

## **Probabilidad Total**

El **teorema de probabilidad total** nos permite calcular la probabilidad de un evento $A$ dividiendo el espacio muestral en varios sub-espacios, y sumando las probabilidades de los eventos condicionados. Este teorema se utiliza cuando no tenemos una forma directa de calcular $P(A)$, pero podemos dividir el espacio muestral en partes m√°s manejables.

### **F√≥rmula General**

La probabilidad total de un evento $A$ se puede calcular como la suma de las probabilidades condicionadas de $A$ dado diferentes eventos $B_1, B_2, \dots, B_n$, cada uno con su propia probabilidad $P(B_i)$:

$$
P(A) = \sum_{i=1}^{n} P(A \mid B_i) \cdot P(B_i)
$$

Donde:
- $P(A \mid B_i)$ es la probabilidad de $A$ dado que ocurri√≥ el evento $B_i$.
- $P(B_i)$ es la probabilidad de que ocurra el evento $B_i$.

### **Aplicaciones**

El teorema de probabilidad total es √∫til cuando tenemos varios caminos diferentes para que ocurra un evento, y cada camino tiene su propia probabilidad. Algunos ejemplos incluyen:
- **Problemas de transiciones de estados**: En un sistema donde se pueden ocurrir varios tipos de transiciones.
- **Problemas de muestreo estratificado**: Como calcular la probabilidad total de obtener una muestra en diferentes grupos de una poblaci√≥n.

###Ejemplo de Probabilidad Conjunta
Supongamos que tenemos una urna con bolas rojas, azules y verdes. Queremos calcular la probabilidad conjunta de sacar una bola roja y luego una bola azul, sin reemplazo.


```python
#Ejemplo de Probabilidad Conjunta
#Supongamos que tenemos una urna con bolas rojas, azules y verdes.
#Queremos calcular la probabilidad conjunta de sacar una bola roja y luego una bola azul, sin reemplazo.

# N√∫mero total de bolas
total_bolas = 10

# N√∫mero de bolas de cada color
rojas = 4
azules = 3
verdes = 3

# Probabilidad de sacar una bola roja
P_roja = rojas / total_bolas

# Probabilidad de sacar una bola azul, dado que ya sacamos una bola roja
P_azul_dada_roja = azules / (total_bolas - 1)

# Probabilidad conjunta de sacar una bola roja y luego una bola azul
P_conjunta = P_roja * P_azul_dada_roja
print(f"Probabilidad conjunta de sacar una bola roja y luego una bola azul: {P_conjunta:.4f}")

# Datos iniciales
colores = ['Rojas', 'Azules', 'Verdes']
cantidad_inicial = [rojas, azules, verdes]
cantidad_despues = [rojas - 1, azules, verdes]  # Se elimina una bola roja

# Configuraci√≥n de la gr√°fica
fig, ax = plt.subplots(figsize=(10, 6))

# Barras para antes y despu√©s
ax.bar(colores, cantidad_inicial, color=['#FF6347', '#4682B4', '#3CB371'], alpha=0.6, label='Antes de la extracci√≥n')
ax.bar(colores, cantidad_despues, color=['#B22222', '#1E3A8A', '#2E8B57'], alpha=0.6, label='Despu√©s de extraer una bola roja')

# Agregar etiquetas sobre las barras
for i in range(len(colores)):
    ax.text(i - 0.1, cantidad_inicial[i] + 0.1, f'{cantidad_inicial[i]}', color='black', fontsize=12)
    ax.text(i - 0.1, cantidad_despues[i] + 0.1, f'{cantidad_despues[i]}', color='black', fontsize=12)

# Etiquetas y t√≠tulo
ax.set_ylabel("Cantidad de bolas")
ax.set_title("Cambio en la Composici√≥n de la Urna tras la Primera Extracci√≥n")
ax.legend()

# Mostrar la grilla
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar gr√°fico
plt.show()
```

### Probabilidad Total
Supongamos que un experimento tiene tres posibles resultados: $B_1$, $B_2$ y $B_3$. Queremos calcular la probabilidad de un evento $A$ dado estos tres eventos posibles.

```python
import matplotlib.pyplot as plt

# Probabilidades de los eventos B1, B2 y B3
probabilidades_B = [0.3, 0.6, 0.1]
eventos = ["B1", "B2", "B3"]

# Probabilidades condicionadas de A dado cada evento Bi
probabilidades_A_dado_B = [0.7, 0.3, 0.8]

# C√°lculo de la contribuci√≥n de cada evento a la probabilidad total de A
contribuciones_A = [probabilidades_B[i] * probabilidades_A_dado_B[i] for i in range(len(eventos))]

# Probabilidad total de A
probabilidad_total_A = sum(contribuciones_A)

# Crear gr√°fico de barras
fig, ax = plt.subplots(figsize=(10, 6))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Barra con contribuciones
bars = ax.bar(eventos, contribuciones_A, color=colors, alpha=0.8)

# Etiquetas sobre las barras
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.005, f'{height:.4f}', ha='center', va='bottom', fontsize=12)

# L√≠nea horizontal indicando la probabilidad total de A
ax.axhline(y=probabilidad_total_A, color='black', linestyle='--', label=f'P(A) = {probabilidad_total_A:.4f}')

# Agregar texto explicativo
ax.text(1, probabilidad_total_A + 0.01, f"Probabilidad Total de A", color='black', fontsize=12)

# Etiquetas y configuraci√≥n
ax.set_ylabel("Contribuci√≥n a P(A)")
ax.set_xlabel("Eventos B")
ax.set_title("Contribuci√≥n de cada Evento B a la Probabilidad Total de A")
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar gr√°fico
plt.tight_layout()
plt.show()
```
