## *Probabilidad Laplaciana*

### 1.1.1 Definición de la Probabilidad Laplaciana

La **probabilidad Laplaciana**, también llamada **probabilidad clásica**, se basa en el principio de **equiprobabilidad**. Se define matemáticamente como:

$$P(A) = \frac{\text{Casos favorables}}{\text{Casos posibles}}
$$

donde:

- \( P(A) \) representa la probabilidad del evento \( A \).
- **"Casos favorables"** son los resultados dentro del espacio muestral que satisfacen \( A \).
- **"Casos posibles"** es el número total de resultados en el espacio muestral \( \Omega \).

Este modelo de probabilidad es válido únicamente cuando **todos los resultados son igualmente probables**. Es decir, si se lanza un dado justo, **cada cara tiene la misma probabilidad de salir**.

### 1.1.2  Teorema Central del Espacio de Equiprobabilidad
El Teorema Central de la Probabilidad Laplaciana establece que si un espacio muestral Ω tiene 𝑁 elementos equiprobables, y un evento 𝐴 tiene 𝑀 de esos elementos, entonces la probabilidad de que ocurra 𝐴 está dada por:

$$𝑃 (𝐴) = \frac{∣𝐴∣}{∣Ω∣}$$
donde:

∣𝐴∣ es el número de elementos que componen el evento 𝐴.

∣ Ω ∣ es el número de elementos del espacio muestral.

Este teorema es la base fundamental de la probabilidad clásica y se usa en experimentos como lanzamiento de monedas, dados y cartas.

### 1.1.3 Demostración del Teorema
Sabemos que en un experimento con equiprobabilidad, todos los casos posibles tienen igual probabilidad de ocurrir. Sea
Ω el espacio muestral con 𝑁 resultados posibles, cada uno con probabilidad:

$$𝑃(𝜔_𝑖)=\frac{1}{𝑁}, \hspace{0.5cm} ∀  𝜔_𝑖 ∈ Ω$$
Si el evento 𝐴 está compuesto por 𝑀 elementos, entonces la probabilidad de ocurrencia de 𝐴 es la suma de las probabilidades de sus elementos individuales:
$$ 𝑃 (𝐴)=∑_{𝜔_𝑖 ∈𝐴}𝑃(𝜔_𝑖)$$
Como cada 𝑃(𝜔_𝑖) es igual a $\frac{1}{𝑁}$, tenemos:
$$
𝑃(𝐴)=∑_{𝜔_𝑖 ∈𝐴}\frac{1}{𝑁}$$
Dado que hay 𝑀 elementos en 𝐴:
$$
𝑃(𝐴)=𝑀⋅\frac{1}{𝑁}=𝑀$$

### 1.2.1 Axiomas de Kolmogorov: Fundamento de la Probabilidad Moderna

La teoría de la probabilidad, tal como la conocemos hoy, se basa en la **formalización matemática** de **Andrei Kolmogorov** en 1933. Él estableció tres axiomas fundamentales que definen cualquier medida de probabilidad sobre un espacio muestral. Estos axiomas son la base sobre la que se construyen conceptos más avanzados como la probabilidad condicional, independencia y distribuciones probabilísticas.

### 1.2.2 Axioma de No Negatividad

La probabilidad de cualquier evento $A$ en el espacio muestral $\Omega$  debe ser mayor o igual a cero:

$$
P(A) \geq 0, \quad \forall A \subseteq \Omega
$$

- **Importancia**: La probabilidad es una medida cuantitativa de incertidumbre. Nunca puede ser negativa, pues representa la proporción de ocurrencia de un evento.
  
- **Aplicación**: Si lanzamos un dado justo, la probabilidad de obtener cualquier número válido está entre 0 y 1. La probabilidad de obtener un número menor que 1 es 0 (evento imposible).


### 1.2.3 Axioma de Normalización

La probabilidad del espacio muestral completo debe ser igual a 1:

$$
P(\Omega) = 1
$$

- **Importancia**: Este axioma establece que algún resultado del espacio muestral debe ocurrir. Es decir, en cualquier experimento hay certeza de que al menos un resultado se manifiesta.
  
- **Aplicación**: En un lanzamiento de una moneda, el espacio muestral es $ \Omega = \{ \text{cara}, \text{cruz} \} $, y sabemos que seguro obtendremos una de esas opciones, por lo que:

$$
P(\text{cara}) + P(\text{cruz}) = 1
$$


### 1.2.4 Axioma de Aditividad (para eventos mutuamente excluyentes)

Si dos eventos $ A $ y $ B $ no tienen resultados en común, entonces:

$$
P(A \cup B) = P(A) + P(B)
$$

- **Importancia**: Si dos eventos no pueden ocurrir simultáneamente, la probabilidad de que ocurra uno u otro es simplemente la suma de sus probabilidades individuales.
  
- **Aplicación**: Supongamos que sacamos una carta de un mazo estándar de 52 cartas. La probabilidad de sacar un As es:

$$
P(A) = \frac{4}{52}
$$

La probabilidad de sacar una Reina es:

$$
P(B) = \frac{4}{52}
$$

Como no podemos sacar un As y una Reina a la vez, aplicamos la aditividad:

$$
P(A \cup B) = P(A) + P(B) = \frac{4}{52} + \frac{4}{52} = \frac{8}{52}
$$

```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
from itertools import product
from collections import Counter
```

```python
# Espacio muestral
espacio_muestral = [1, 2, 3, 4, 5, 6]

# Casos favorables: números pares
favorables = [x for x in espacio_muestral if x % 2 == 0]

# Probabilidad clásica
p_par = len(favorables) / len(espacio_muestral)
print(f"La probabilidad de obtener un número par es: {p_par:.2f}")
```

```python
# Espacio muestral: todos los pares (d1, d2)
espacio = list(itertools.product(range(1, 7), repeat=2))
n_total = len(espacio)

# Evento: suma igual a 7
suma_frecuencias = [sum(p) for p in espacio]
frecuencias = Counter(suma_frecuencias)
n_favorables = frecuencias[7]

# Probabilidad clásica
P = n_favorables / n_total
print(f"Probabilidad de suma 7: {P:.4f}")

# Crear histograma
plt.hist(suma_frecuencias, bins=range(2, 14), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.axvline(7, color='red', linestyle='--', label='Suma 7')
plt.title('Distribución de Sumas de Dos Dados')
plt.xlabel('Suma')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar probabilidad en el gráfico
plt.text(7, max(plt.hist(suma_frecuencias, bins=range(2, 14), align='left', rwidth=0.8)[0]),
         f'P(7) = {P:.4f}', ha='center', va='bottom', color='red')

plt.show()
```

## Probabilidad Geométrica

La **probabilidad geométrica** se aplica cuando el espacio muestral es un conjunto continuo y la probabilidad de un evento se determina en función de la "medida" del conjunto favorable en relación al tamaño total del espacio muestral. Es decir, la probabilidad de un evento es proporcional al tamaño del conjunto favorable dentro del espacio muestral.

### Fórmula General

Sea $A$ un evento en un espacio muestral continuo $\Omega$, la probabilidad de $A$ se calcula como:

$$
P(A) = \frac{\text{medida}(A)}{\text{medida}(\Omega)}
$$

Donde:
- $A$ es el conjunto de eventos favorables,
- $\Omega$ es el espacio muestral total,
- $\text{medida}(A)$ representa el "tamaño" del conjunto favorable, que puede ser una longitud, área, volumen, etc.
- $\text{medida}(\Omega)$ es el "tamaño" total del espacio muestral.


### Aplicaciones

Este tipo de probabilidad es útil en muchos campos, como la **geometría** y la **física**, cuando se modelan fenómenos aleatorios en espacios continuos. Además, es especialmente relevante cuando trabajamos con **modelos de áreas** o **volúmenes**.

```python
# Contexto:
# El problema de la aguja de Buffon consiste en estimar el valor de pi mediante la probabilidad de que
# una aguja de longitud L, lanzada aleatoriamente sobre líneas paralelas separadas por una distancia d,
# cruce alguna de las líneas. Cuantas más veces lanzamos la aguja, mejor será la estimación de pi.
# Este código simula el problema de la aguja de Buffon para estimar el valor de pi.
# Parámetros

import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
L = 1.0  # Longitud de la aguja
d = 1.0  # Distancia entre las líneas paralelas
num_trials = 100000  # Número total de lanzamientos
step = 1000  # Número de lanzamientos por paso para la estimación de pi

# Listas para almacenar los valores de pi estimados y los lanzamientos
pi_estimates = []
trial_steps = []

# Contador para cuántas agujas cruzan una línea
crossed_lines = 0

# Simulación de lanzamientos de la aguja y estimación de pi en pasos
for trial in range(1, num_trials + 1):
    # Posición del centro de la aguja
    center_position = np.random.uniform(0, d / 2)

    # Ángulo aleatorio entre 0 y pi (ángulo en radianes)
    angle = np.random.uniform(0, np.pi)

    # Distancia desde el centro de la aguja hasta la línea más cercana
    distance_to_nearest_line = L / 2 * np.sin(angle)

    # Verificamos si la aguja cruza una línea
    if center_position <= distance_to_nearest_line:
        crossed_lines += 1

    # Estimación de pi en pasos
    if trial % step == 0 and crossed_lines > 0:
        pi_estimate = (2 * L * trial) / (crossed_lines * d)
        pi_estimates.append(pi_estimate)
        trial_steps.append(trial)

print(f"Estimación de pi: {pi_estimate}")

# Plot de la estimación de pi a lo largo de los lanzamientos
plt.figure(figsize=(10, 6))
plt.plot(trial_steps, pi_estimates, label="Estimación de π", color="blue")
plt.axhline(y=np.pi, color="red", linestyle="--", label="Valor real de π")
plt.xlabel('Número de lanzamientos')
plt.ylabel('Estimación de π')
plt.title('Convergencia de la estimación de π con el problema de la aguja de Buffon')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Problema de la Zona de Intersección de dos Círculos: Con ayuda de la probabilidad geometrica y segmentos circulares,
#se calculará la zona compartida entre dos círculos en función de sus radios y la distancia entre sus centros.
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
r = 1  # Radio de los círculos
n = 22000  # Número de simulaciones
intersecciones = 0

# Función para simular la posición de los círculos
def circulos_interseccion():
    # Posiciones aleatorias del centro de los círculos
    x1, y1 = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
    x2, y2 = np.random.uniform(-5, 5), np.random.uniform(-5, 5)

    # Calcular la distancia entre los centros
    distancia_centros = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    # Si la distancia entre los centros es menor o igual a 2*r, los círculos se intersectan
    return distancia_centros <= 2 * r

# Simulación
for _ in range(n):
    if circulos_interseccion():
        intersecciones += 1

# Estimación de la probabilidad
probabilidad = intersecciones / n
print(f"Probabilidad de intersección de los círculos: {probabilidad:.4f}")

# Visualización de una simulación aleatoria
x1, y1 = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
x2, y2 = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
fig, ax = plt.subplots()

# Círculo 1
circle1 = plt.Circle((x1, y1), r, color='blue', alpha=0.5)
ax.add_artist(circle1)

# Círculo 2
circle2 = plt.Circle((x2, y2), r, color='red', alpha=0.5)
ax.add_artist(circle2)

# Configuración de la gráfica
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_aspect('equal', 'box')
ax.set_title('Simulación de Intersección de Círculos')
plt.show()
```

## **Probabilidad Frecuentista**

La **probabilidad frecuentista** se basa en la observación de la frecuencia relativa de los eventos en experimentos repetidos. Según este enfoque, la probabilidad de un evento se define como el **límite** de la frecuencia relativa de su ocurrencia conforme el número de experimentos aumenta infinitamente.

Este enfoque es especialmente útil cuando no conocemos la probabilidad exacta de un evento, pero podemos **realizar experimentos** y **contabilizar** cuántas veces ocurre el evento en cuestión.

### **Fórmula General**

Supongamos que repetimos un experimento aleatorio $n$ veces, y el evento $A$ ocurre $f(A)$ veces. Entonces, la **probabilidad frecuentista** de que ocurra el evento $A$ se calcula como:

$$
P(A) = \lim_{n \to \infty} \frac{f(A)}{n}
$$

Donde:
- $f(A)$ es el número de veces que ocurre el evento $A$ en $n$ experimentos,
- $n$ es el número total de experimentos.

A medida que $n$ se hace muy grande, la **frecuencia relativa** de la ocurrencia del evento $A$ se aproxima a su **probabilidad verdadera**.

### **Conceptos Clave**

1. **Experimentos Aleatorios**: Un **experimento aleatorio** es un proceso cuyo resultado no es predecible con certeza, pero se conoce el conjunto de todos los resultados posibles (espacio muestral).

2. **Frecuencia Relativa**: La frecuencia relativa de un evento es la **proporción de veces** que ocurre en relación con el número total de experimentos realizados. Se calcula como:

$$
\text{Frecuencia Relativa} = \frac{f(A)}{n}
$$

3. **Ley de los Grandes Números**: Según esta ley, conforme el número de experimentos aumenta, la frecuencia relativa de un evento se estabiliza y se aproxima a su probabilidad teórica. Formalmente, se expresa como:

$$
\lim_{n \to \infty} \frac{f(A)}{n} = P(A)
$$

4. **Independencia de Experimentos**: En el enfoque frecuentista, los experimentos son **independientes**

```python
#Ejemplo del Volado

# Fijar la semilla para reproducibilidad
np.random.seed(42)

# Definir parámetros
num_lanzamientos = 10_000
opciones = ['sol', 'aguila']

# Simulación de lanzamientos
resultados = np.random.choice(opciones, size=num_lanzamientos)

# Cálculo de frecuencia relativa acumulada
frecuencia_sol = np.cumsum(resultados == 'sol') / np.arange(1, num_lanzamientos + 1)

# Configuración de la gráfica
plt.figure(figsize=(10, 5))
plt.plot(frecuencia_sol, label='Frecuencia relativa de "sol"', color='blue')
plt.axhline(0.5, color='red', linestyle='--', label='Valor teórico: 0.5')

# Personalización de etiquetas y título
plt.title('Probabilidad Frecuentista: Convergencia de Frecuencia Relativa')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Frecuencia relativa')
plt.legend()
plt.grid(True)

# Mostrar gráfica
plt.show()
```

```python
# Simulación de un proceso de Poisson
np.random.seed(42)
n = 10000
lambda_poisson = 3  # Tasa de ocurrencias
resultados_poisson = np.random.poisson(lambda_poisson, n)

# Frecuencia relativa de los eventos
frecuencia_poisson = np.cumsum(resultados_poisson == 3) / np.arange(1, n + 1)

# Graficar
plt.figure(figsize=(10, 5))
plt.plot(frecuencia_poisson, label=f'Frecuencia relativa de ocurrencias de 3', color='green')
plt.axhline(1/np.exp(lambda_poisson), color='red', linestyle='--', label=f'Valor teórico para {lambda_poisson}')
plt.title('Probabilidad Frecuentista: Proceso de Poisson')
plt.xlabel('Número de experimentos')
plt.ylabel('Frecuencia relativa')
plt.grid(True)
plt.legend()
plt.show()
```

## **Probabilidad Conjunta**

La **probabilidad conjunta** se refiere a la probabilidad de que ocurran dos (o más) eventos al mismo tiempo. Es decir, es la probabilidad de que los eventos $A$ y $B$ ocurran simultáneamente, y se den en conjunto. Se denota como $P(A \cap B)$, donde "$\cap$" indica la intersección de los eventos.

### **Fórmula General**

La probabilidad conjunta de dos eventos $A$ y $B$ es:

$$
P(A \cap B) = P(A \mid B) \cdot P(B)
$$

Donde:
- $P(A \mid B)$ es la **probabilidad condicionada** de que ocurra el evento $A$, dado que ya ocurrió el evento $B$.
- $P(B)$ es la probabilidad de que ocurra el evento $B$.

### **Condición de Independencia**

Si los eventos $A$ y $B$ son **independientes**, entonces la probabilidad conjunta se simplifica a:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

Esto significa que la ocurrencia de uno de los eventos no afecta la probabilidad de que ocurra el otro evento.

### **Aplicaciones**

La probabilidad conjunta se utiliza en una variedad de situaciones, como:
- **Problemas de urnas**: ¿Cuál es la probabilidad de sacar una bola roja y una bola azul de una urna?
- **Encuestas y estudios de mercado**: Calcular la probabilidad de que una persona tenga ciertas características simultáneamente (por ejemplo, que sea hombre y fumador).

## **Probabilidad Total**

El **teorema de probabilidad total** nos permite calcular la probabilidad de un evento $A$ dividiendo el espacio muestral en varios sub-espacios, y sumando las probabilidades de los eventos condicionados. Este teorema se utiliza cuando no tenemos una forma directa de calcular $P(A)$, pero podemos dividir el espacio muestral en partes más manejables.

### **Fórmula General**

La probabilidad total de un evento $A$ se puede calcular como la suma de las probabilidades condicionadas de $A$ dado diferentes eventos $B_1, B_2, \dots, B_n$, cada uno con su propia probabilidad $P(B_i)$:

$$
P(A) = \sum_{i=1}^{n} P(A \mid B_i) \cdot P(B_i)
$$

Donde:
- $P(A \mid B_i)$ es la probabilidad de $A$ dado que ocurrió el evento $B_i$.
- $P(B_i)$ es la probabilidad de que ocurra el evento $B_i$.

### **Aplicaciones**

El teorema de probabilidad total es útil cuando tenemos varios caminos diferentes para que ocurra un evento, y cada camino tiene su propia probabilidad. Algunos ejemplos incluyen:
- **Problemas de transiciones de estados**: En un sistema donde se pueden ocurrir varios tipos de transiciones.
- **Problemas de muestreo estratificado**: Como calcular la probabilidad total de obtener una muestra en diferentes grupos de una población.

Supongamos que tenemos una urna con bolas rojas, azules y verdes. Queremos calcular la probabilidad conjunta de sacar una bola roja y luego una bola azul, sin reemplazo.


```python
#Ejemplo de Probabilidad Conjunta
#Supongamos que tenemos una urna con bolas rojas, azules y verdes.
#Queremos calcular la probabilidad conjunta de sacar una bola roja y luego una bola azul, sin reemplazo.

# Número total de bolas
total_bolas = 10

# Número de bolas de cada color
rojas = 4
azules = 3
verdes = 3

# Probabilidad de sacar una bola roja
P_roja = rojas / total_bolas

# Probabilidad de sacar una bola azul, dado que ya sacamos una bola roja
P_azul_dada_roja = azules / (total_bolas - 1)

# Probabilidad conjunta de sacar una bola roja y luego una bola azul
P_conjunta = P_roja * P_azul_dada_roja
print(f"Probabilidad conjunta de sacar una bola roja y luego una bola azul: {P_conjunta:.4f}")

# Datos iniciales
colores = ['Rojas', 'Azules', 'Verdes']
cantidad_inicial = [rojas, azules, verdes]
cantidad_despues = [rojas - 1, azules, verdes]  # Se elimina una bola roja

# Configuración de la gráfica
fig, ax = plt.subplots(figsize=(10, 6))

# Barras para antes y después
ax.bar(colores, cantidad_inicial, color=['#FF6347', '#4682B4', '#3CB371'], alpha=0.6, label='Antes de la extracción')
ax.bar(colores, cantidad_despues, color=['#B22222', '#1E3A8A', '#2E8B57'], alpha=0.6, label='Después de extraer una bola roja')

# Agregar etiquetas sobre las barras
for i in range(len(colores)):
    ax.text(i - 0.1, cantidad_inicial[i] + 0.1, f'{cantidad_inicial[i]}', color='black', fontsize=12)
    ax.text(i - 0.1, cantidad_despues[i] + 0.1, f'{cantidad_despues[i]}', color='black', fontsize=12)

# Etiquetas y título
ax.set_ylabel("Cantidad de bolas")
ax.set_title("Cambio en la Composición de la Urna tras la Primera Extracción")
ax.legend()

# Mostrar la grilla
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar gráfico
plt.show()
```

### Probabilidad Total
Supongamos que un experimento tiene tres posibles resultados: $B_1$, $B_2$ y $B_3$. Queremos calcular la probabilidad de un evento $A$ dado estos tres eventos posibles.

```python
import matplotlib.pyplot as plt

# Probabilidades de los eventos B1, B2 y B3
probabilidades_B = [0.3, 0.6, 0.1]
eventos = ["B1", "B2", "B3"]

# Probabilidades condicionadas de A dado cada evento Bi
probabilidades_A_dado_B = [0.7, 0.3, 0.8]

# Cálculo de la contribución de cada evento a la probabilidad total de A
contribuciones_A = [probabilidades_B[i] * probabilidades_A_dado_B[i] for i in range(len(eventos))]

# Probabilidad total de A
probabilidad_total_A = sum(contribuciones_A)

# Crear gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Barra con contribuciones
bars = ax.bar(eventos, contribuciones_A, color=colors, alpha=0.8)

# Etiquetas sobre las barras
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.005, f'{height:.4f}', ha='center', va='bottom', fontsize=12)

# Línea horizontal indicando la probabilidad total de A
ax.axhline(y=probabilidad_total_A, color='black', linestyle='--', label=f'P(A) = {probabilidad_total_A:.4f}')

# Agregar texto explicativo
ax.text(1, probabilidad_total_A + 0.01, f"Probabilidad Total de A", color='black', fontsize=12)

# Etiquetas y configuración
ax.set_ylabel("Contribución a P(A)")
ax.set_xlabel("Eventos B")
ax.set_title("Contribución de cada Evento B a la Probabilidad Total de A")
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```
